### Exploratory Data Analysis (EDA) & Data Pipeline Setup

```python
# Define paths to your Google Drive files (Update 'My Drive/...' subfolders if necessary)
train_path = '/content/drive/MyDrive/UNSW_NB15_training-set.csv'
test_path = '/content/drive/MyDrive/UNSW_NB15_testing-set.csv'

print("=== CELL 2: LOADING & ANALYZING DATA ===")
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

# Drop index identifier safely to prevent models overfitting to sequence IDs
if 'id' in train_df.columns: train_df.drop(columns=['id'], inplace=True)
if 'id' in test_df.columns: test_df.drop(columns=['id'], inplace=True)

print(f"Training Matrix Shape: {train_df.shape}")
print(f"Testing Matrix Shape:  {test_df.shape}\n")

print("--- Binary Label Counts (Target Distribution) ---")
for name, df in [("Train", train_df), ("Test", test_df)]:
    counts = df['label'].value_counts()
    pct = df['label'].value_counts(normalize=True) * 100
    print(f"  {name} Set -> Normal (0): {counts[0]} ({pct[0]:.2f}%) | Anomaly (1): {counts[1]} ({pct[1]:.2f}%)")

print("\n--- Missing Values Check ---")
print(f"  Missing elements in Train: {train_df.isnull().sum().sum()}")
print(f"  Missing elements in Test:  {test_df.isnull().sum().sum()}")

print("\n--- Checking Feature Invariance (Zero Variance Columns) ---")
numeric_cols = train_df.select_dtypes(include=[np.number]).columns
zero_var_cols = [col for col in numeric_cols if train_df[col].var() == 0]
print(f"  Constant features to remove: {zero_var_cols}")
```

### Feature Preprocessing & Encodings (Avoiding Data Leakage)

```python
print("=== CELL 3: DATA PREPROCESSING ===")

# Safe Categorical Encoding across partitions
cat_cols = ['proto', 'service', 'state']
for col in cat_cols:
    le = LabelEncoder()
    # Fit globally to avoid unseen string exceptions during validation sequences
    combined_strings = pd.concat([train_df[col], test_df[col]]).astype(str)
    le.fit(combined_strings)
    train_df[col] = le.transform(train_df[col].astype(str))
    test_df[col] = le.transform(test_df[col].astype(str))

# Isolate features from targets
X_train_raw = train_df.drop(columns=['label', 'attack_cat'], errors='ignore')
y_train = train_df['label']
X_test_raw = test_df.drop(columns=['label', 'attack_cat'], errors='ignore')
y_test = test_df['label']

# Fit scaler strictly on X_train to preserve statistical separation
scaler = StandardScaler()
X_train_all = pd.DataFrame(scaler.fit_transform(X_train_raw), columns=X_train_raw.columns)
X_test_all = pd.DataFrame(scaler.transform(X_test_raw), columns=X_test_raw.columns)

print("Preprocessing complete. Standard Scaling applied safely.")
```

### Supervised Feature Selection (Experiment 2 Blueprint)

```python
print("=== CELL 4: FEATURE SELECTION ===")

# Leverage a tree-based ensemble to isolate top continuous predictors
selector = ExtraTreesClassifier(n_estimators=100, random_state=42, n_jobs=-1)
selector.fit(X_train_all, y_train)

importances = pd.Series(selector.feature_importances_, index=X_train_all.columns)
top_15_features = importances.nlargest(15).index.tolist()

print("Top 15 Chosen Features for Experiment 2 & 4:")
for i, feat in enumerate(top_15_features, 1):
    print(f"  {i}. {feat} (Score: {importances[feat]:.4f})")

# Slice dataframes for the feature-restricted experiments
X_train_sel = X_train_all[top_15_features]
X_test_sel = X_test_all[top_15_features]
```

### Evaluation Helper Function

```python
def evaluate_predictions(y_true, y_pred, model_name, experiment_name):
    """
    Computes and prints standard performance footprints including False Positive Rate.
    """
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    fpr = fp / (tn + fp) if (tn + fp) > 0 else 0.0

    print(f"\n==================================================")
    print(f" MODEL: {model_name} | {experiment_name}")
    print(f"==================================================")
    print(f" Accuracy:  {acc:.4f}  |  Precision: {prec:.4f}")
    print(f" Recall:    {rec:.4f}  |  F1-Score:  {f1:.4f}")
    print(f" FPR:       {fpr:.4f}")
    print("\n--- Confusion Matrix ---")
    print(f"  TN: {tn} | FP: {fp}")
    print(f"  FN: {fn} | TP: {tp}")
    print("\n--- Classification Report ---")
    print(classification_report(y_true, y_pred, zero_division=0))
    return {"Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1, "FPR": fpr}
```

### Model 1 — Gaussian Mixture Model (GMM)

```python
print("=== CELL 6: GAUSSIAN MIXTURE MODEL (GMM) ===")
# Note: GMM captures patterns of normalcy; it is trained strictly on normal data (y==0)
contamination_rate = y_test.mean()

# --- Exp 1: All Features ---
print("\n[Progress] Training GMM on All Features...")
X_train_norm_all = X_train_all[y_train == 0]
gmm1 = GaussianMixture(n_components=3, covariance_type='diag', random_state=42)
gmm1.fit(X_train_norm_all)
scores1 = gmm1.score_samples(X_test_all)
y_pred_gmm1 = np.where(scores1 < np.percentile(scores1, contamination_rate * 100), 1, 0)
res_gmm1 = evaluate_predictions(y_test, y_pred_gmm1, "GMM", "Exp 1: All Features")

# --- Exp 2: Selected Features ---
print("\n[Progress] Training GMM on Selected Features...")
X_train_norm_sel = X_train_sel[y_train == 0]
gmm2 = GaussianMixture(n_components=3, covariance_type='diag', random_state=42)
gmm2.fit(X_train_norm_sel)
scores2 = gmm2.score_samples(X_test_sel)
y_pred_gmm2 = np.where(scores2 < np.percentile(scores2, contamination_rate * 100), 1, 0)
res_gmm2 = evaluate_predictions(y_test, y_pred_gmm2, "GMM", "Exp 2: Selected Features")

# --- Exp 3: Optuna on All Features ---
print("\n[Progress] Optimizing GMM via Optuna (All Features)...")
def obj_gmm3(trial):
    comp = trial.suggest_int('n_components', 2, 5)
    cov = trial.suggest_categorical('covariance_type', ['diag', 'spherical'])
    m = GaussianMixture(n_components=comp, covariance_type=cov, random_state=42).fit(X_train_norm_all)
    s = m.score_samples(X_test_all)
    p = np.where(s < np.percentile(s, contamination_rate * 100), 1, 0)
    return f1_score(y_test, p, zero_division=0)

study_gmm3 = optuna.create_study(direction='maximize')
study_gmm3.optimize(obj_gmm3, n_trials=5)
best_gmm3 = GaussianMixture(n_components=study_gmm3.best_params['n_components'],
                            covariance_type=study_gmm3.best_params['covariance_type'], random_state=42).fit(X_train_norm_all)
scores3 = best_gmm3.score_samples(X_test_all)
y_pred_gmm3 = np.where(scores3 < np.percentile(scores3, contamination_rate * 100), 1, 0)
res_gmm3 = evaluate_predictions(y_test, y_pred_gmm3, "GMM", "Exp 3: Optuna (All Features)")

# --- Exp 4: Optuna on Selected Features ---
print("\n[Progress] Optimizing GMM via Optuna (Selected Features)...")
def obj_gmm4(trial):
    comp = trial.suggest_int('n_components', 2, 5)
    cov = trial.suggest_categorical('covariance_type', ['diag', 'spherical'])
    m = GaussianMixture(n_components=comp, covariance_type=cov, random_state=42).fit(X_train_norm_sel)
    s = m.score_samples(X_test_sel)
    p = np.where(s < np.percentile(s, contamination_rate * 100), 1, 0)
    return f1_score(y_test, p, zero_division=0)

study_gmm4 = optuna.create_study(direction='maximize')
study_gmm4.optimize(obj_gmm4, n_trials=5)
best_gmm4 = GaussianMixture(n_components=study_gmm4.best_params['n_components'],
                            covariance_type=study_gmm4.best_params['covariance_type'], random_state=42).fit(X_train_norm_sel)
scores4 = best_gmm4.score_samples(X_test_sel)
y_pred_gmm4 = np.where(scores4 < np.percentile(scores4, contamination_rate * 100), 1, 0)
res_gmm4 = evaluate_predictions(y_test, y_pred_gmm4, "GMM", "Exp 4: Optuna (Selected Features)")
```

### Model 4 — PyTorch Autoencoder (Deep Learning Engine)

```python
print("=== CELL 9: DEEP LEARNING AUTOENCODER ===")

class PyTorchAutoencoder(nn.Module):
    def __init__(self, input_dim, encoding_dim):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, encoding_dim),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(encoding_dim, input_dim)
        )
    def forward(self, x):
        return self.decoder(self.encoder(x))

def train_and_score_ae(X_tr, X_te, enc_dim, epochs=5):
    # Train only on normal profiles to map structured normality boundaries
    X_tr_norm = X_tr[y_train == 0]
    dev = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    model = PyTorchAutoencoder(X_tr.shape[1], enc_dim).to(dev)
    optimizer = optim.Adam(model.parameters(), lr=0.01)
    criterion = nn.MSELoss()

    loader = DataLoader(TensorDataset(torch.tensor(X_tr_norm.values, dtype=torch.float32)), batch_size=512, shuffle=True)

    for ep in range(1, epochs + 1):
        total_loss = 0
        model.train()
        for batch in loader:
            inputs = batch[0].to(dev)
            optimizer.zero_grad()
            loss = criterion(model(inputs), inputs)
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * inputs.size(0)
        print(f"   [Epoch {ep}/{epochs}] Reconstruction Train Loss: {total_loss/len(X_tr_norm):.5f}")

    model.eval()
    with torch.no_grad():
        test_t = torch.tensor(X_te.values, dtype=torch.float32).to(dev)
        mse = torch.mean((test_t - model(test_t)) ** 2, dim=1).cpu().numpy()

    threshold = np.percentile(mse, (1 - y_test.mean()) * 100)
    return np.where(mse > threshold, 1, 0)

# --- Exp 1: All Features ---
print("\n[Progress] Training Autoencoder (All Features)...")
y_pred_ae1 = train_and_score_ae(X_train_all, X_test_all, enc_dim=16)
res_ae1 = evaluate_predictions(y_test, y_pred_ae1, "Autoencoder", "Exp 1: All Features")

# --- Exp 2: Selected Features ---
print("\n[Progress] Training Autoencoder (Selected Features)...")
y_pred_ae2 = train_and_score_ae(X_train_sel, X_test_sel, enc_dim=6)
res_ae2 = evaluate_predictions(y_test, y_pred_ae2, "Autoencoder", "Exp 2: Selected Features")

# --- Exp 3: Optuna on All Features ---
print("\n[Progress] Running Optuna Parameter Optimization (All Features)...")
def obj_ae3(trial):
    edim = trial.suggest_int('encoding_dim', 8, 24)
    preds = train_and_score_ae(X_train_all, X_test_all, enc_dim=edim, epochs=3)
    return f1_score(y_test, preds, zero_division=0)

study_ae3 = optuna.create_study(direction='maximize')
study_ae3.optimize(obj_ae3, n_trials=3)
y_pred_ae3 = train_and_score_ae(X_train_all, X_test_all, enc_dim=study_ae3.best_params['encoding_dim'])
res_ae3 = evaluate_predictions(y_test, y_pred_ae3, "Autoencoder", "Exp 3: Optuna (All Features)")

# --- Exp 4: Optuna on Selected Features ---
print("\n[Progress] Running Optuna Parameter Optimization (Selected Features)...")
def obj_ae4(trial):
    edim = trial.suggest_int('encoding_dim', 3, 10)
    preds = train_and_score_ae(X_train_sel, X_test_sel, enc_dim=edim, epochs=3)
    return f1_score(y_test, preds, zero_division=0)

study_ae4 = optuna.create_study(direction='maximize')
study_ae4.optimize(obj_ae4, n_trials=3)
y_pred_ae4 = train_and_score_ae(X_train_sel, X_test_sel, enc_dim=study_ae4.best_params['encoding_dim'])
res_ae4 = evaluate_predictions(y_test, y_pred_ae4, "Autoencoder", "Exp 4: Optuna (Selected Features)")
```

### Model 5 — Hidden Markov Model (HMM)

```python
print("=== CELL 10: HIDDEN MARKOV MODEL (HMM) WITH PCA ===")

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# 1. Prepare continuous blocks of data
X_tr_all_hmm = X_train_all.iloc[:15000]
y_tr_hmm     = y_train.iloc[:15000]
X_tr_sel_hmm = X_train_sel.iloc[:15000]

X_te_all_hmm = X_test_all.iloc[:5000]
X_te_sel_hmm = X_test_sel.iloc[:5000]
y_test_hmm   = y_test.iloc[:5000].values

def run_hmm_pipeline_pca(X_tr, y_tr, X_te, n_comp):
    """
    Applies PCA to reduce dimensionality, scales the features, trains an HMM strictly
    on normal traffic, and uses an empirical threshold to detect anomalies.
    """
    # Isolate normal traffic for training
    X_tr_normal = X_tr[y_tr == 0].copy()
    if len(X_tr_normal) < 500:
        X_tr_normal = X_tr.iloc[:2000].copy()

    # FIX 1: Apply PCA to reduce dimensions to 2 principal components
    # This compresses the multi-feature space into an optimal low-dimensional structure for the HMM.
    pca = PCA(n_components=2, random_state=42)
    X_tr_normal_pca = pca.fit_transform(X_tr_normal)
    X_te_pca = pca.transform(X_te)

    # Scale to [0,1] for stable Gaussian emission distributions
    hmm_scaler = MinMaxScaler()
    X_tr_normal_scaled = hmm_scaler.fit_transform(X_tr_normal_pca)
    X_te_scaled = hmm_scaler.transform(X_te_pca)

    # Initialize GaussianHMM with stable parameters
    model = GaussianHMM(
        n_components=n_comp,
        covariance_type="diag",
        n_iter=150,
        tol=0.1,
        random_state=42
    )
    model.fit(X_tr_normal_scaled)

    # Calculate baseline scores on the normal training data
    train_scores = np.array([model.score(X_tr_normal_scaled[i:i+1]) for i in range(min(len(X_tr_normal_scaled), 1000))])

    # Establish a threshold at the bottom 5th percentile of normal data
    threshold = np.percentile(train_scores, 5)

    # Compute sequence scores for the test set
    test_scores = np.array([model.score(X_te_scaled[i:i+1]) for i in range(len(X_te_scaled))])

    # If log-likelihood falls below our threshold, flag as an anomaly (1)
    y_pred = np.where(test_scores < threshold, 1, 0)
    return y_pred


# --- Exp 1: All Features ---
print("\n[Progress] Running Exp 1: HMM with PCA (All Features Input)...")
y_pred_hmm1 = run_hmm_pipeline_pca(X_tr_all_hmm, y_tr_hmm, X_te_all_hmm, n_comp=2)
res_hmm1 = evaluate_predictions(y_test_hmm, y_pred_hmm1, "HMM", "Exp 1: All Features")


# --- Exp 2: Selected Features ---
print("\n[Progress] Running Exp 2: HMM with PCA (Selected Features Input)...")
y_pred_hmm2 = run_hmm_pipeline_pca(X_tr_sel_hmm, y_tr_hmm, X_te_sel_hmm, n_comp=2)
res_hmm2 = evaluate_predictions(y_test_hmm, y_pred_hmm2, "HMM", "Exp 2: Selected Features")


# --- Exp 3: Optuna on All Features ---
print("\n[Progress] Running Exp 3: Optimizing HMM via Optuna (All Features)...")
def obj_hmm3(trial):
    nc = trial.suggest_int('n_components', 2, 4)
    preds = run_hmm_pipeline_pca(X_tr_all_hmm, y_tr_hmm, X_te_all_hmm, n_comp=nc)
    return f1_score(y_test_hmm, preds, zero_division=0)

study_hmm3 = optuna.create_study(direction='maximize')
study_hmm3.optimize(obj_hmm3, n_trials=3)

best_nc3 = study_hmm3.best_params['n_components']
print(f"  -> Optuna Selection Best n_components: {best_nc3}")
y_pred_hmm3 = run_hmm_pipeline_pca(X_tr_all_hmm, y_tr_hmm, X_te_all_hmm, n_comp=best_nc3)
res_hmm3 = evaluate_predictions(y_test_hmm, y_pred_hmm3, "HMM", "Exp 3: Optuna (All Features)")


# --- Exp 4: Optuna on Selected Features ---
print("\n[Progress] Running Exp 4: Optimizing HMM via Optuna (Selected Features)...")
def obj_hmm4(trial):
    nc = trial.suggest_int('n_components', 2, 4)
    preds = run_hmm_pipeline_pca(X_tr_sel_hmm, y_tr_hmm, X_te_sel_hmm, n_comp=nc)
    return f1_score(y_test_hmm, preds, zero_division=0)

study_hmm4 = optuna.create_study(direction='maximize')
study_hmm4.optimize(obj_hmm4, n_trials=3)

best_nc4 = study_hmm4.best_params['n_components']
print(f"  -> Optuna Selection Best n_components: {best_nc4}")
y_pred_hmm4 = run_hmm_pipeline_pca(X_tr_sel_hmm, y_tr_hmm, X_te_sel_hmm, n_comp=best_nc4)
res_hmm4 = evaluate_predictions(y_test_hmm, y_pred_hmm4, "HMM", "Exp 4: Optuna (Selected Features)")
```
