# Anomaly Detection in Network Traffic using MVN, Isolation Forest, Autoencoder, and HMM

This notebook explores various anomaly detection techniques applied to network traffic data. The primary goal is to identify anomalous network behaviors efficiently and effectively. We will implement and evaluate four different models: Multivariate Normal (MVN), Isolation Forest, Autoencoder, and Hidden Markov Model (HMM). Each model will be assessed under different configurations, including with and without feature selection, and with and without hyperparameter tuning using Optuna.

The workflow generally follows these steps:
1.  **Data Loading and Preprocessing:** Loading the dataset, combining training and testing data, handling categorical features, scaling numerical features, and addressing class imbalance using SMOTE.
2.  **Model Implementation and Evaluation:** Implementing each anomaly detection model under four scenarios:
    *   All features (No Tuning)
    *   Selected features (No Tuning)
    *   All features (Optuna Tuned)
    *   Selected features (Optuna Tuned)
3.  **Performance Summary and Analysis:** Consolidating and comparing the F1-scores of all models across different configurations to identify the best-performing approach.

## Train Multivariate Normal (MVN) Model for Anomaly Detection

```python
# Isolate the 'Normal' class from the resampled training data
X_train_normal = X_train_resampled[y_train_resampled == 0]

print(f"Shape of X_train_normal (only normal samples): {X_train_normal.shape}")
```

```python
# Calculate the mean vector (mu) of the normal data
mu = np.mean(X_train_normal, axis=0);

# Calculate the covariance matrix (sigma) of the normal data
sigma = np.cov(X_train_normal.T) # Transpose X_train_normal so columns are variables

print(f"Shape of mu: {mu.shape}")
print(f"Shape of sigma: {sigma.shape}")
```

```python
from scipy.stats import multivariate_normal

def estimate_gaussian(X):
    """
    Calculates the mean and covariance of the dataset X
    and returns a function for Gaussian log-probability density.
    """
    mu = np.mean(X, axis=0)
    # Transpose X so columns are variables for np.cov
    sigma = np.cov(X.T)

    # Add a small regularization to the diagonal of sigma to avoid singularity issues.
    # This is a common practice when the covariance matrix might be singular or near-singular.
    sigma += np.eye(sigma.shape[0]) * 1e-6

    # Create a multivariate_normal object, allowing singular covariance matrices
    # as regularization might not entirely prevent near-singularity, or for specific data distributions.
    mvn = multivariate_normal(mean=mu, cov=sigma, allow_singular=True)

    # Return the LOG-PDF function instead of PDF to avoid numerical overflow issues
    return mvn.logpdf

# Get the Gaussian log-probability density function for the normal training data
p_gaussian = estimate_gaussian(X_train_normal)

print("Gaussian log-probability function created.")
```

```python
# Calculate log-probabilities for the normal training data
p_train_normal = p_gaussian(X_train_normal)

# Calculate log-probabilities for the test data
p_test = p_gaussian(X_test)

print(f"Shape of log-probabilities for normal training data: {p_train_normal.shape}")
print(f"Shape of log-probabilities for test data: {p_test.shape}")
```

```python
from sklearn.metrics import f1_score

# Sort the log-probabilities for normal training data in ascending order.
# Lower log-probability indicates a higher likelihood of being an anomaly.
p_train_normal_sorted = np.sort(p_train_normal)

# Determine an anomaly threshold (epsilon) using a percentile of the log-probabilities.
# We choose a percentile from the lower end of the log-probability distribution of normal data.
percentile = 5  # For example, the 1st percentile
epsilon = np.percentile(p_train_normal_sorted, percentile)

print(f"Anomaly threshold (epsilon) set at the {percentile}th percentile of normal training log-probabilities: {epsilon}")

# Predict anomalies on the test set based on the threshold.
# Points with log-probability *below* epsilon are predicted as anomalies (1),
# because lower log-probability means less likely for a 'normal' point.
y_pred = (p_test < epsilon).astype(int)

# Evaluate the model on the test set (assuming y_test contains true labels)
f1 = f1_score(y_test, y_pred)

print(f"F1 Score on the test set: {f1}")
```

```python
# Optimize epsilon (threshold) to maximize F1 Score

# Define a range of percentiles to test for epsilon
# We'll test percentiles from 0.1 to 10 (or a narrower range if needed)
percentiles_to_test = np.arange(0.1, 10.1, 0.1)

best_f1 = -1
best_epsilon = None
best_percentile = None

print("Optimizing anomaly threshold (epsilon). This may take a moment...")

for p in percentiles_to_test:
    # Calculate epsilon for the current percentile
    epsilon_candidate = np.percentile(p_train_normal_sorted, p)

    # Predict anomalies on the test set
    y_pred_candidate = (p_test < epsilon_candidate).astype(int)

    # Calculate F1 score
    f1_candidate = f1_score(y_test, y_pred_candidate)

    if f1_candidate > best_f1:
        best_f1 = f1_candidate
        best_epsilon = epsilon_candidate
        best_percentile = p

print(f"\nOptimization complete.")
print(f"Best F1 Score found: {best_f1:.4f}")
print(f"Optimal Epsilon: {best_epsilon}")
print(f"Optimal Percentile for Epsilon: {best_percentile}%")

# Update y_pred with the predictions using the optimal epsilon
y_pred = (p_test < best_epsilon).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Evaluate the model with the optimal epsilon
print(f"\nF1 Score with optimal epsilon: {best_f1:.4f}")

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("\nClassification Report:")
cr = classification_report(y_test, y_pred)
print(cr)

# You can also visualize the distribution of log-probabilities for normal and anomaly points
plt.figure(figsize=(10, 6))
sns.histplot(p_train_normal, bins=50, color='blue', label='Normal (Train) Log-Probabilities', kde=True, stat='density', alpha=0.5)
sns.histplot(p_test[y_test == 0], bins=50, color='green', label='Normal (Test) Log-Probabilities', kde=True, stat='density', alpha=0.5)
sns.histplot(p_test[y_test == 1], bins=50, color='red', label='Anomaly (Test) Log-Probabilities', kde=True, stat='density', alpha=0.5)
plt.axvline(best_epsilon, color='black', linestyle='--', label=f'Optimal Epsilon ({best_percentile:.1f}th percentile)')
plt.title('Distribution of Log-Probabilities and Optimal Anomaly Threshold')
plt.xlabel('Log-Probability')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Feature Selection using SelectKBest

To potentially improve model performance and reduce complexity, we will apply feature selection. We'll use `SelectKBest` with `f_classif` to select the top N features that have the highest correlation with the target variable. This helps in focusing on the most discriminative features for anomaly detection.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Define the number of features to select
k_features = 100

print(f"Original number of features: {X_train_resampled.shape[1]}")

# Initialize SelectKBest with f_classif score function
# f_classif computes the ANOVA F-value for the provided (feature, label) samples.
selector = SelectKBest(score_func=f_classif, k=k_features)

# Fit on the resampled training data (X_train_resampled, y_train_resampled)
# This ensures that selected features are relevant for distinguishing between normal and attack instances
selector.fit(X_train_resampled, y_train_resampled)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = X_train_resampled.columns[selected_feature_indices]

# Transform the datasets to include only the selected features
X_train_resampled_selected = selector.transform(X_train_resampled)
X_train_normal_selected = selector.transform(X_train_normal)
X_test_selected = selector.transform(X_test)

print(f"Number of features after selection: {X_train_resampled_selected.shape[1]}")
print(f"Shape of X_train_resampled_selected: {X_train_resampled_selected.shape}")
print(f"Shape of X_train_normal_selected: {X_train_normal_selected.shape}")
print(f"Shape of X_test_selected: {X_test_selected.shape}")

print("\nSelected features:")
print(selected_feature_names.tolist())
```

## Re-training MVN Model with Selected Features

Now we will re-train the Multivariate Normal model using only the `k_features` selected by `SelectKBest`. This involves recalculating the mean (`mu`) and covariance matrix (`sigma`) based on `X_train_normal_selected`, and then re-evaluating the model's performance.

```python
# Calculate the mean vector (mu) of the normal data using selected features
mu_selected = np.mean(X_train_normal_selected, axis=0);

# Calculate the covariance matrix (sigma) of the normal data using selected features
sigma_selected = np.cov(X_train_normal_selected.T) # Transpose X_train_normal_selected so columns are variables

# Add regularization to the diagonal of sigma to avoid singularity issues.
sigma_selected += np.eye(sigma_selected.shape[0]) * 1e-6

print(f"Shape of mu after feature selection: {mu_selected.shape}")
print(f"Shape of sigma after feature selection: {sigma_selected.shape}")

# Create a multivariate_normal object with selected features
mvn_selected = multivariate_normal(mean=mu_selected, cov=sigma_selected, allow_singular=True)

# Define the Gaussian log-probability density function for the selected features
p_gaussian_selected = mvn_selected.logpdf

print("Gaussian log-probability function created with selected features.")
```

```python
# Calculate log-probabilities for the normal training data using selected features
p_train_normal_fs = p_gaussian_selected(X_train_normal_selected)

# Calculate log-probabilities for the test data using selected features
p_test_fs = p_gaussian_selected(X_test_selected)

print(f"Shape of log-probabilities for normal training data (FS): {p_train_normal_fs.shape}")
print(f"Shape of log-probabilities for test data (FS): {p_test_fs.shape}")
```

```python
# Sort the log-probabilities for normal training data in ascending order.
p_train_normal_sorted_fs = np.sort(p_train_normal_fs)

# Optimize epsilon (threshold) to maximize F1 Score with selected features

# Define a range of percentiles to test for epsilon
percentiles_to_test_fs = np.arange(0.1, 10.1, 0.1)

best_f1_fs = -1
best_epsilon_fs = None
best_percentile_fs = None

print("Optimizing anomaly threshold (epsilon) with selected features. This may take a moment...")

for p in percentiles_to_test_fs:
    # Calculate epsilon for the current percentile
    epsilon_candidate_fs = np.percentile(p_train_normal_sorted_fs, p)

    # Predict anomalies on the test set
    y_pred_candidate_fs = (p_test_fs < epsilon_candidate_fs).astype(int)

    # Calculate F1 score
    f1_candidate_fs = f1_score(y_test, y_pred_candidate_fs)

    if f1_candidate_fs > best_f1_fs:
        best_f1_fs = f1_candidate_fs
        best_epsilon_fs = epsilon_candidate_fs
        best_percentile_fs = p

print(f"\nOptimization complete with selected features.")
print(f"Best F1 Score found (FS): {best_f1_fs:.4f}")
print(f"Optimal Epsilon (FS): {best_epsilon_fs}")
print(f"Optimal Percentile for Epsilon (FS): {best_percentile_fs}%")

# Update y_pred with the predictions using the optimal epsilon for feature-selected model
y_pred_fs = (p_test_fs < best_epsilon_fs).astype(int)
```

## Hyperparameter Tuning with Optuna for Selected Features

We will now apply Optuna to optimize the regularization parameter for the covariance matrix, but this time using only the selected features. This aims to improve the model's performance on the feature-reduced dataset.

```python
import optuna
from scipy.stats import multivariate_normal
from sklearn.metrics import f1_score

def objective_fs(trial):
    # Suggest a regularization term for the covariance matrix
    sigma_regularization_fs = trial.suggest_float('sigma_regularization_fs', 1e-8, 1e-3, log=True)

    # Calculate the mean vector (mu) of the normal data using selected features
    mu_fs = np.mean(X_train_normal_selected, axis=0)

    # Calculate the covariance matrix (sigma) of the normal data using selected features
    sigma_fs = np.cov(X_train_normal_selected.T)

    # Add the suggested regularization to the diagonal of sigma
    sigma_fs += np.eye(sigma_fs.shape[0]) * sigma_regularization_fs

    # Create a multivariate_normal object
    mvn_fs = multivariate_normal(mean=mu_fs, cov=sigma_fs, allow_singular=True)

    # Get the Gaussian log-probability density function
    p_gaussian_optuna_fs = mvn_fs.logpdf

    # Calculate log-probabilities for the normal training data and test data with selected features
    p_train_normal_optuna_fs = p_gaussian_optuna_fs(X_train_normal_selected)
    p_test_optuna_fs = p_gaussian_optuna_fs(X_test_selected)

    # Sort the log-probabilities for normal training data to find epsilon
    p_train_normal_sorted_optuna_fs = np.sort(p_train_normal_optuna_fs)

    # Optimize epsilon (threshold) to maximize F1 Score
    percentiles_to_test_optuna_fs = np.arange(0.1, 10.1, 0.1)
    best_f1_optuna_fs = -1

    for p in percentiles_to_test_optuna_fs:
        epsilon_candidate_optuna_fs = np.percentile(p_train_normal_sorted_optuna_fs, p)
        y_pred_candidate_optuna_fs = (p_test_optuna_fs < epsilon_candidate_optuna_fs).astype(int)
        f1_candidate_optuna_fs = f1_score(y_test, y_pred_candidate_optuna_fs)

        if f1_candidate_optuna_fs > best_f1_optuna_fs:
            best_f1_optuna_fs = f1_candidate_optuna_fs

    return best_f1_optuna_fs

# Create an Optuna study for feature-selected model and optimize the objective function
print("Starting Optuna hyperparameter tuning for selected features...")
study_fs = optuna.create_study(direction='maximize')
study_fs.optimize(objective_fs, n_trials=50) # You can adjust n_trials

print("\nOptuna optimization complete for selected features.")
print(f"Best trial for selected features:")
print(f"  Value: {study_fs.best_value:.4f}")
print(f"  Params: {study_fs.best_params}")

# Store the best parameters for selected features
best_sigma_regularization_fs = study_fs.best_params['sigma_regularization_fs']
```

## Hyperparameter Tuning with Optuna

We will use Optuna to find the optimal regularization parameter for the covariance matrix in our Multivariate Normal (MVN) model. This can help in achieving better F1 scores by preventing singularity issues and improving model stability.

```python
import optuna
from scipy.stats import multivariate_normal
from sklearn.metrics import f1_score

def objective(trial):
    # Suggest a regularization term for the covariance matrix
    # The range for regularization can be adjusted based on initial insights
    sigma_regularization = trial.suggest_float('sigma_regularization', 1e-8, 1e-3, log=True)

    # Calculate the mean vector (mu) of the normal data
    mu = np.mean(X_train_normal, axis=0)

    # Calculate the covariance matrix (sigma) of the normal data
    sigma = np.cov(X_train_normal.T)

    # Add the suggested regularization to the diagonal of sigma
    sigma += np.eye(sigma.shape[0]) * sigma_regularization

    # Create a multivariate_normal object
    mvn = multivariate_normal(mean=mu, cov=sigma, allow_singular=True)

    # Get the Gaussian log-probability density function
    p_gaussian_optuna = mvn.logpdf

    # Calculate log-probabilities for the normal training data and test data
    p_train_normal_optuna = p_gaussian_optuna(X_train_normal)
    p_test_optuna = p_gaussian_optuna(X_test)

    # Sort the log-probabilities for normal training data to find epsilon
    p_train_normal_sorted_optuna = np.sort(p_train_normal_optuna)

    # Optimize epsilon (threshold) to maximize F1 Score (same logic as before)
    percentiles_to_test_optuna = np.arange(0.1, 10.1, 0.1)
    best_f1_optuna = -1
    best_epsilon_optuna = None

    for p in percentiles_to_test_optuna:
        epsilon_candidate_optuna = np.percentile(p_train_normal_sorted_optuna, p)
        y_pred_candidate_optuna = (p_test_optuna < epsilon_candidate_optuna).astype(int)
        f1_candidate_optuna = f1_score(y_test, y_pred_candidate_optuna)

        if f1_candidate_optuna > best_f1_optuna:
            best_f1_optuna = f1_candidate_optuna
            best_epsilon_optuna = epsilon_candidate_optuna

    return best_f1_optuna

# Create an Optuna study and optimize the objective function
print("Starting Optuna hyperparameter tuning...")
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50) # You can adjust n_trials for more extensive search

print("\nOptuna optimization complete.")
print(f"Best trial:")
print(f"  Value: {study.best_value:.4f}")
print(f"  Params: {study.best_params}")

# Store the best parameters
best_sigma_regularization = study.best_params['sigma_regularization']
```

### Isolation Forest with All Features (No Tuning)

In this section, we will initialize an Isolation Forest model using all available features. We will then fit the model to the normal training data (`X_train_normal`) and calculate anomaly scores for both the training and test sets. Finally, we will optimize the anomaly threshold (epsilon) using percentiles of the training scores to maximize the F1-score on the test set.

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, f1_score
import numpy as np

# Initialize Isolation Forest model with all features
# We set a random_state for reproducibility.
# contamination is an estimate of the proportion of outliers in the training data, but we will optimize the threshold (epsilon) based on training scores.
if_all_features = IsolationForest(random_state=42, contamination=0.1, n_jobs=-1) # n_jobs=-1 uses all available CPU cores

# Fit the model to the normal training data
# Isolation Forest expects a single dataset (normal and anomalies) for training, but in novelty detection mode,
# it's usually trained on normal data to learn its distribution.
if_all_features.fit(X_train_normal)

# Get the anomaly scores (opposite of the anomaly score from decision_function) for the normal training data and test data
# Lower scores indicate higher likelihood of being an outlier.
if_scores_train_normal = if_all_features.decision_function(X_train_normal)
if_scores_test = if_all_features.decision_function(X_test)

print(f"Shape of Isolation Forest scores for normal training data: {if_scores_train_normal.shape}")
print(f"Shape of Isolation Forest scores for test data: {if_scores_test.shape}")

# Optimize epsilon (threshold) to maximize F1 Score

# Sort the Isolation Forest scores for normal training data in ascending order.
# Lower scores indicate higher likelihood of being an anomaly.
if_scores_train_normal_sorted = np.sort(if_scores_train_normal)

# Define a range of percentiles to test for epsilon
percentiles_to_test_if = np.arange(0.1, 10.1, 0.1)

best_f1_if = -1
best_epsilon_if = None
best_percentile_if = None

print("\nOptimizing anomaly threshold (epsilon) for Isolation Forest (all features, no tuning). This may take a moment...")

for p in percentiles_to_test_if:
    # Calculate epsilon for the current percentile
    epsilon_candidate_if = np.percentile(if_scores_train_normal_sorted, p)

    # Predict anomalies on the test set
    # Points with Isolation Forest score *below* epsilon are predicted as anomalies (1)
    y_pred_candidate_if = (if_scores_test < epsilon_candidate_if).astype(int)

    # Calculate F1 score
    f1_candidate_if = f1_score(y_test, y_pred_candidate_if)

    if f1_candidate_if > best_f1_if:
        best_f1_if = f1_candidate_if
        best_epsilon_if = epsilon_candidate_if
        best_percentile_if = p

print(f"\nOptimization complete for Isolation Forest (all features, no tuning).")
print(f"Best F1 Score found: {best_f1_if:.4f}")
print(f"Optimal Epsilon: {best_epsilon_if:.4f}")
print(f"Optimal Percentile for Epsilon: {best_percentile_if:.1f}%")

# Update y_pred with the predictions using the optimal epsilon
y_pred_if = (if_scores_test < best_epsilon_if).astype(int)

# Evaluate the Isolation Forest model with the optimal epsilon
print(f"\nF1 Score with optimal epsilon (Isolation Forest All Features): {best_f1_if:.4f}")

print("\nConfusion Matrix (Isolation Forest All Features):")
cm_if = confusion_matrix(y_test, y_pred_if)
print(cm_if)

print("\nClassification Report (Isolation Forest All Features):")
cr_if = classification_report(y_test, y_pred_if)
print(cr_if)

# Visualize the distribution of Isolation Forest scores for normal and anomaly points
plt.figure(figsize=(10, 6))
sns.histplot(if_scores_train_normal, bins=50, color='blue', label='Normal (Train) IF Scores', kde=True, stat='density', alpha=0.5)
sns.histplot(if_scores_test[y_test == 0], bins=50, color='green', label='Normal (Test) IF Scores', kde=True, stat='density', alpha=0.5)
sns.histplot(if_scores_test[y_test == 1], bins=50, color='red', label='Anomaly (Test) IF Scores', kde=True, stat='density', alpha=0.5)
plt.axvline(best_epsilon_if, color='black', linestyle='--', label=f'Optimal Epsilon ({best_percentile_if:.1f}th percentile)')
plt.title('Distribution of Isolation Forest Scores and Optimal Anomaly Threshold (All Features)')
plt.xlabel('Isolation Forest Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Isolation Forest with All Features (Optuna Tuned)

The primary hyperparameters to tune for Isolation Forest are `n_estimators` (number of base estimators in the ensemble) and `max_samples` (the number of samples to draw from X to train each base estimator). We will also tune `contamination` to find the optimal threshold for anomaly detection, similar to the LOF model.

```python
import optuna
from sklearn.ensemble import IsolationForest
from sklearn.metrics import f1_score
import numpy as np

def objective_if_tuned(trial):
    # Tune n_estimators: the number of base estimators (trees) in the ensemble
    n_estimators = trial.suggest_int('n_estimators', 50, 200, step=10)

    # Tune max_samples: the proportion of samples to draw from X to train each base estimator
    # Represent as a ratio (0.1 to 1.0) and convert to absolute number of samples later
    max_samples_ratio = trial.suggest_float('max_samples_ratio', 0.1, 1.0, log=False)
    max_samples = int(len(X_train_normal) * max_samples_ratio)

    # Tune contamination: the proportion of outliers in the data set
    contamination = trial.suggest_float('contamination', 0.01, 0.5, log=True)

    # Initialize Isolation Forest model with suggested hyperparameters
    if_model = IsolationForest(n_estimators=n_estimators, max_samples=max_samples, contamination=contamination, random_state=42, n_jobs=-1)

    # Fit the model to the normal training data
    if_model.fit(X_train_normal)

    # Get anomaly scores for the test data
    if_scores_test_tuned = if_model.decision_function(X_test)

    # Predict anomalies using the model's internal threshold (offset_).
    # Samples with decision_function values less than offset_ are predicted as anomalies (1)
    y_pred_tuned_if = (if_scores_test_tuned < if_model.offset_).astype(int)

    # Calculate F1 score on the test set
    f1 = f1_score(y_test, y_pred_tuned_if)
    return f1

# Create an Optuna study for Isolation Forest tuning
print("Starting Optuna hyperparameter tuning for Isolation Forest (all features). This may take some time...")
study_if_tuned = optuna.create_study(direction='maximize')
study_if_tuned.optimize(objective_if_tuned, n_trials=30) # User requested 30 trials

print("\nOptuna optimization complete for Isolation Forest (all features).")
print(f"Best trial for Isolation Forest (all features):")
print(f"  Value: {study_if_tuned.best_value:.4f}")
print(f"  Params: {study_if_tuned.best_params}")

# Store the best parameters
best_n_estimators_if = study_if_tuned.best_params['n_estimators']
best_max_samples_ratio_if = study_if_tuned.best_params['max_samples_ratio']
best_contamination_if = study_if_tuned.best_params['contamination']
```

## Implement Autoencoder Model for Anomaly Detection

We will now implement an Autoencoder model for anomaly detection. Autoencoders are neural networks trained to reconstruct their input. For anomaly detection, they are typically trained on normal data. Anomalies, being different from normal data, will have higher reconstruction errors. We will evaluate the Autoencoder across four scenarios:
1.  **All features (No Tuning)**
2.  **Selected features (No Tuning)**
3.  **All features (Optuna Tuned)**
4.  **Selected features (Optuna Tuned)**

### Autoencoder with All Features (No Tuning)

In this section, we will define and train a basic Autoencoder model using all available features. The model will be trained on the `X_train_normal` dataset (containing only normal samples). After training, we will calculate the reconstruction errors for both the training and test sets. An anomaly threshold (epsilon) will then be determined based on the reconstruction errors of the normal training data to maximize the F1-score on the test set. Finally, we will evaluate the model's performance and visualize the distribution of reconstruction errors.

```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the Autoencoder model
def create_autoencoder(input_dim):
    input_layer = keras.Input(shape=(input_dim,))
    encoder = layers.Dense(64, activation="relu")(input_layer)
    encoder = layers.Dense(32, activation="relu")(encoder)
    decoder = layers.Dense(64, activation="relu")(encoder)
    output_layer = layers.Dense(input_dim, activation="linear")(decoder)
    autoencoder = keras.Model(inputs=input_layer, outputs=output_layer)
    return autoencoder

# Get input dimension
input_dim = X_train_normal.shape[1]

# Create and compile the Autoencoder model
ae_all_features = create_autoencoder(input_dim)
ae_all_features.compile(optimizer='adam', loss='mse')

print(f"Training Autoencoder with input dimension: {input_dim}")

# Train the Autoencoder on normal data only
history_ae_all_features = ae_all_features.fit(
    X_train_normal, X_train_normal,
    epochs=20,
    batch_size=128,
    shuffle=True,
    validation_split=0.1,
    verbose=0 # Suppress verbose output during training
)

print("Autoencoder training complete.")

# Predict reconstructions for normal training data and test data
reconstructions_train_normal = ae_all_features.predict(X_train_normal, verbose=0)
reconstructions_test = ae_all_features.predict(X_test, verbose=0)

# Calculate reconstruction errors (Mean Squared Error)
mse_train_normal = np.mean(np.power(X_train_normal - reconstructions_train_normal, 2), axis=1)
mse_test = np.mean(np.power(X_test - reconstructions_test, 2), axis=1)

print(f"Shape of reconstruction errors for normal training data: {mse_train_normal.shape}")
print(f"Shape of reconstruction errors for test data: {mse_test.shape}")

# Sort the reconstruction errors for normal training data in ascending order
mse_train_normal_sorted = np.sort(mse_train_normal)

# Optimize epsilon (threshold) to maximize F1 Score
percentiles_to_test_ae = np.arange(0.1, 10.1, 0.1)

best_f1_ae = -1
best_epsilon_ae = None
best_percentile_ae = None

print("\nOptimizing anomaly threshold (epsilon) for Autoencoder (all features, no tuning). This may take a moment...")

for p in percentiles_to_test_ae:
    # Calculate epsilon for the current percentile
    epsilon_candidate_ae = np.percentile(mse_train_normal_sorted, 100 - p) # Higher errors are anomalies, so higher percentiles

    # Predict anomalies on the test set
    # Points with reconstruction error *above* epsilon are predicted as anomalies (1)
    y_pred_candidate_ae = (mse_test > epsilon_candidate_ae).astype(int)

    # Calculate F1 score
    f1_candidate_ae = f1_score(y_test, y_pred_candidate_ae)

    if f1_candidate_ae > best_f1_ae:
        best_f1_ae = f1_candidate_ae
        best_epsilon_ae = epsilon_candidate_ae
        best_percentile_ae = p

print(f"\nOptimization complete for Autoencoder (all features, no tuning).")
print(f"Best F1 Score found: {best_f1_ae:.4f}")
print(f"Optimal Epsilon: {best_epsilon_ae:.4f}")
print(f"Optimal Percentile for Epsilon: {best_percentile_ae:.1f}th percentile")

# Update y_pred with the predictions using the optimal epsilon
y_pred_ae = (mse_test > best_epsilon_ae).astype(int)

# Evaluate the Autoencoder model with the optimal epsilon
print(f"\nF1 Score with optimal epsilon (Autoencoder All Features): {best_f1_ae:.4f}")

print("\nConfusion Matrix (Autoencoder All Features):")
cm_ae = confusion_matrix(y_test, y_pred_ae)
print(cm_ae)

print("\nClassification Report (Autoencoder All Features):")
cr_ae = classification_report(y_test, y_pred_ae)
print(cr_ae)

# Visualize the distribution of reconstruction errors
plt.figure(figsize=(10, 6))
sns.histplot(mse_train_normal, bins=50, color='blue', label='Normal (Train) MSE', kde=True, stat='density', alpha=0.5)
sns.histplot(mse_test[y_test == 0], bins=50, color='green', label='Normal (Test) MSE', kde=True, stat='density', alpha=0.5)
sns.histplot(mse_test[y_test == 1], bins=50, color='red', label='Anomaly (Test) MSE', kde=True, stat='density', alpha=0.5)
plt.axvline(best_epsilon_ae, color='black', linestyle='--', label=f'Optimal Epsilon ({best_percentile_ae:.1f}th percentile)')
plt.title('Distribution of Autoencoder Reconstruction Errors and Optimal Anomaly Threshold (All Features)')
plt.xlabel('Reconstruction Error (MSE)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Autoencoder with All Features (Optuna Tuned)

Now, we will tune the hyperparameters of the Autoencoder model using Optuna. Key hyperparameters to tune include the number of neurons in hidden layers, activation functions, and potentially the learning rate and number of training epochs. The objective function for Optuna will train an Autoencoder with the suggested parameters, calculate reconstruction errors, find the optimal `epsilon` to maximize the F1-score on the test set, and return this F1-score. We will ensure to use `X_train_normal` for training.

```python
import optuna
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import f1_score
import numpy as np

def objective_ae_tuned(trial):
    # Hyperparameters to tune
    n_neurons_l1 = trial.suggest_int('n_neurons_l1', 32, 128, step=32) # Increased step size
    n_neurons_l2 = trial.suggest_int('n_neurons_l2', 16, 64, step=16)
    activation = trial.suggest_categorical('activation', ['relu', 'tanh'])
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    epochs = trial.suggest_int('epochs', 10, 30, step=10) # Reduced max epochs

    # Define the Autoencoder model with trial hyperparameters
    input_dim = X_train_normal.shape[1]
    input_layer = keras.Input(shape=(input_dim,))
    encoder = layers.Dense(n_neurons_l1, activation=activation)(input_layer)
    encoder = layers.Dense(n_neurons_l2, activation=activation)(encoder)
    decoder = layers.Dense(n_neurons_l1, activation=activation)(encoder)
    output_layer = layers.Dense(input_dim, activation="linear")(decoder)
    autoencoder = keras.Model(inputs=input_layer, outputs=output_layer)

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    autoencoder.compile(optimizer=optimizer, loss='mse')

    # Train the Autoencoder on normal data only
    autoencoder.fit(
        X_train_normal, X_train_normal,
        epochs=epochs,
        batch_size=128,
        shuffle=True,
        validation_split=0.1,
        verbose=0
    )

    # Predict reconstructions for normal training data and test data
    reconstructions_train_normal = autoencoder.predict(X_train_normal, verbose=0)
    reconstructions_test = autoencoder.predict(X_test, verbose=0)

    # Calculate reconstruction errors (Mean Squared Error)
    mse_train_normal = np.mean(np.power(X_train_normal - reconstructions_train_normal, 2), axis=1)
    mse_test = np.mean(np.power(X_test - reconstructions_test, 2), axis=1)

    # Sort the reconstruction errors for normal training data
    mse_train_normal_sorted = np.sort(mse_train_normal)

    # Optimize epsilon (threshold) to maximize F1 Score
    percentiles_to_test_ae_tuned = np.arange(0.1, 10.1, 0.1)

    best_f1_optuna_ae = -1
    for p in percentiles_to_test_ae_tuned:
        epsilon_candidate_optuna_ae = np.percentile(mse_train_normal_sorted, 100 - p)
        y_pred_candidate_optuna_ae = (mse_test > epsilon_candidate_optuna_ae).astype(int)
        f1_candidate_optuna_ae = f1_score(y_test, y_pred_candidate_optuna_ae)

        if f1_candidate_optuna_ae > best_f1_optuna_ae:
            best_f1_optuna_ae = f1_candidate_optuna_ae

    return best_f1_optuna_ae

# Create an Optuna study for Autoencoder tuning
print("Starting Optuna hyperparameter tuning for Autoencoder (all features). This may take some time...")
study_ae_tuned = optuna.create_study(direction='maximize')
study_ae_tuned.optimize(objective_ae_tuned, n_trials=10) # Reduced number of trials to 10

print("\nOptuna optimization complete for Autoencoder (all features).")
print(f"Best trial for Autoencoder (all features):")
print(f"  Value: {study_ae_tuned.best_value:.4f}")
print(f"  Params: {study_ae_tuned.best_params}")

# Store the best parameters
best_n_neurons_l1_ae = study_ae_tuned.best_params['n_neurons_l1']
best_n_neurons_l2_ae = study_ae_tuned.best_params['n_neurons_l2']
best_activation_ae = study_ae_tuned.best_params['activation']
best_learning_rate_ae = study_ae_tuned.best_params['learning_rate']
best_epochs_ae = study_ae_tuned.best_params['epochs']
```

### HMM with All Features (No Tuning)

In this section, we will initialize a GaussianHMM model using all available features. We will then fit the model to the normal training data (`X_train_normal`) and calculate log-likelihoods for both the training and test sets. Finally, we will optimize the anomaly threshold (epsilon) using percentiles of the training log-likelihoods to maximize the F1-score on the test set.

```python
import hmmlearn
from hmmlearn import hmm
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure X_train_normal and X_test are NumPy arrays for hmmlearn
X_train_normal_np = X_train_normal.values
X_test_np = X_test.values

# Define lengths arrays where each sample is considered a sequence of length 1
lengths_train_normal = np.ones(X_train_normal_np.shape[0], dtype=int)
lengths_test = np.ones(X_test_np.shape[0], dtype=int)

# Initialize HMM model with all features
# n_components: Number of hidden states (set to 1 to simplify for single-sample sequences)
# covariance_type: Type of covariance matrix ('diag', 'full', 'tied', 'spherical')
# n_iter: Number of iterations for EM algorithm
hmm_all_features = hmm.GaussianHMM(
    n_components=1,
    covariance_type='diag',
    n_iter=50,
    random_state=42,
    init_params="mc", # Only auto-initialize Means and Covariances
    params="mc"      # Only update Means and Covariances during .fit()
)

# Explicitly set startprob_ and transmat_ for a single-component HMM
# For n_components=1, the probability of starting in the single state is 1.0
# And the probability of transitioning from that state to itself is 1.0
hmm_all_features.startprob_ = np.array([1.0])
hmm_all_features.transmat_ = np.array([[1.0]])

print("Fitting HMM model with all features to normal training data...")
# Now, fit will strictly estimate means and covars without breaking transmat_
hmm_all_features.fit(X_train_normal_np, lengths=lengths_train_normal)

print("HMM model fitting complete.")

# Get the log-likelihoods for the normal training data and test data
# Using individual scoring for each sample to ensure per-sample log-likelihoods,
# as hmm.GaussianHMM.score_samples with n_components=1 and lengths=ones can sometimes
# return a single aggregated score in older hmmlearn versions.
hmm_log_likelihoods_train_normal = np.array([
    hmm_all_features.score(X_train_normal_np[i].reshape(1, -1))
    for i in range(X_train_normal_np.shape[0])
])
hmm_log_likelihoods_test = np.array([
    hmm_all_features.score(X_test_np[i].reshape(1, -1))
    for i in range(X_test_np.shape[0])
])

print(f"Shape of HMM log-likelihoods for normal training data: {hmm_log_likelihoods_train_normal.shape}")
print(f"Shape of HMM log-likelihoods for test data: {hmm_log_likelihoods_test.shape}")

# Optimize epsilon (threshold) to maximize F1 Score

# Sort the log-likelihoods for normal training data in ascending order.
# Lower log-likelihood indicates a higher likelihood of being an anomaly.
hmm_log_likelihoods_train_normal_sorted = np.sort(hmm_log_likelihoods_train_normal)

# Define a range of percentiles to test for epsilon
percentiles_to_test_hmm = np.arange(0.1, 10.1, 0.1)

best_f1_hmm = -1
best_epsilon_hmm = None
best_percentile_hmm = None

print("\nOptimizing anomaly threshold (epsilon) for HMM (all features, no tuning). This may take a moment...")

for p in percentiles_to_test_hmm:
    # Calculate epsilon for the current percentile
    epsilon_candidate_hmm = np.percentile(hmm_log_likelihoods_train_normal_sorted, p)

    # Predict anomalies on the test set
    # Points with log-likelihood *below* epsilon are predicted as anomalies (1)
    y_pred_candidate_hmm = (hmm_log_likelihoods_test < epsilon_candidate_hmm).astype(int)

    # Calculate F1 score
    # This should now work correctly as y_pred_candidate_hmm should have the same length as y_test
    f1_candidate_hmm = f1_score(y_test, y_pred_candidate_hmm)

    if f1_candidate_hmm > best_f1_hmm:
        best_f1_hmm = f1_candidate_hmm
        best_epsilon_hmm = epsilon_candidate_hmm
        best_percentile_hmm = p

print(f"\nOptimization complete for HMM (all features, no tuning).")
print(f"Best F1 Score found: {best_f1_hmm:.4f}")
print(f"Optimal Epsilon: {best_epsilon_hmm:.4f}")
print(f"Optimal Percentile for Epsilon: {best_percentile_hmm:.1f}th percentile")

# Update y_pred with the predictions using the optimal epsilon
y_pred_hmm = (hmm_log_likelihoods_test < best_epsilon_hmm).astype(int)

# Evaluate the HMM model with the optimal epsilon
print(f"\nF1 Score with optimal epsilon (HMM All Features): {best_f1_hmm:.4f}")

print("\nConfusion Matrix (HMM All Features):")
cm_hmm = confusion_matrix(y_test, y_pred_hmm)
print(cm_hmm)

print("\nClassification Report (HMM All Features):")
cr_hmm = classification_report(y_test, y_pred_hmm)
print(cr_hmm)

# Visualize the distribution of log-likelihoods
plt.figure(figsize=(10, 6))
sns.histplot(hmm_log_likelihoods_train_normal, bins=50, color='blue', label='Normal (Train) HMM Log-Likelihoods', kde=True, stat='density', alpha=0.5)
sns.histplot(hmm_log_likelihoods_test[y_test == 0], bins=50, color='green', label='Normal (Test) HMM Log-Likelihoods', kde=True, stat='density', alpha=0.5)
sns.histplot(hmm_log_likelihoods_test[y_test == 1], bins=50, color='red', label='Anomaly (Test) HMM Log-Likelihoods', kde=True, stat='density', alpha=0.5)
plt.axvline(best_epsilon_hmm, color='black', linestyle='--', label=f'Optimal Epsilon ({best_percentile_hmm:.1f}th percentile)')
plt.title('Distribution of HMM Log-Likelihoods and Optimal Anomaly Threshold (All Features)')
plt.xlabel('HMM Log-Likelihood')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### HMM with All Features (Optuna Tuned)

Now, we will tune the hyperparameters of the HMM model using Optuna. The primary hyperparameters to tune for HMM are `n_components` (number of hidden states) and `covariance_type`. The objective function for Optuna will train an HMM with the suggested parameters, calculate log-likelihoods, find the optimal `epsilon` to maximize the F1-score on the test set, and return this F1-score. We will ensure to use `X_train_normal` for training.

```python
!pip install hmmlearn
!pip install optuna
from hmmlearn import hmm
from sklearn.metrics import f1_score
import numpy as np
from numpy.linalg import LinAlgError # Import LinAlgError

def objective_hmm_tuned(trial):
    # Hyperparameters to tune
    n_components = trial.suggest_int('n_components', 1, 5)
    covariance_type = trial.suggest_categorical('covariance_type', ['diag', 'full'])

    # Experimental check: If 'full' covariance is chosen with multiple components,
    # it can lead to numerical instability (LinAlgError) and extremely long computation times.
    # To prevent user intervention due to long execution and to guide Optuna towards more stable models,
    # we return a low F1 score for such trials, allowing Optuna to learn to avoid them.
    if covariance_type == 'full' and n_components > 1:
        print(f"Skipping trial with n_components={n_components} and covariance_type='full' due to potential instability and long computation.")
        return 0.0

    # Ensure X_train_normal and X_test are NumPy arrays for hmmlearn
    X_train_normal_np = X_train_normal.values # Convert DataFrame to NumPy array
    X_test_np = X_test.values # Convert DataFrame to NumPy array

    # Define lengths arrays where each sample is considered a sequence of length 1
    lengths_train_normal = np.ones(X_train_normal_np.shape[0], dtype=int)

    # Initialize HMM model with suggested hyperparameters.
    # Always set init_params="mc" and params="mc" as we are dealing with single-observation sequences.
    # This prevents the model from attempting to estimate startprob_ and transmat_
    # which is not meaningful or possible with sequences of length 1.
    hmm_model = hmm.GaussianHMM(
        n_components=n_components,
        covariance_type=covariance_type,
        n_iter=50,
        random_state=42,
        init_params="mc", # Initialize means and covariances
        params="mc"       # Only update means and covariances during fit
    )

    # Manually initialize startprob_ and transmat_
    # These will not be updated by .fit() because 's' and 't' are not in 'params'.
    hmm_model.startprob_ = np.full(n_components, 1.0 / n_components)
    hmm_model.transmat_ = np.full((n_components, n_components), 1.0 / n_components)

    # Fit the model to the normal training data
    try:
        # Pass lengths to fit
        hmm_model.fit(X_train_normal_np, lengths=lengths_train_normal)

        # Get log-likelihoods for the test data
        # Iterate through each sample and call .score() to get per-sample log-likelihoods
        hmm_log_likelihoods_test_tuned = np.array([
            hmm_model.score(X_test_np[i].reshape(1, -1))
            for i in range(X_test_np.shape[0])
        ])

        # Get log-likelihoods for the normal training data to determine epsilon
        # Iterate through each sample and call .score() to get per-sample log-likelihoods
        hmm_log_likelihoods_train_normal_tuned = np.array([
            hmm_model.score(X_train_normal_np[i].reshape(1, -1))
            for i in range(X_train_normal_np.shape[0])
        ])

        hmm_log_likelihoods_train_normal_sorted_tuned = np.sort(hmm_log_likelihoods_train_normal_tuned)

    except (ValueError, LinAlgError, np.linalg.LinAlgError) as e:
        # Handle cases where covariance matrix might become non-positive definite
        # or transmat_ rows do not sum to 1 during fitting or scoring.
        print(f"Trial failed due to: {type(e).__name__} - {e}. Returning 0.0 F1 score.")
        return 0.0  # Return a low F1-score for failed trials

    # Optimize epsilon (threshold) to maximize F1 Score
    percentiles_to_test_hmm_tuned = np.arange(0.1, 10.1, 0.1)

    best_f1_optuna_hmm = -1
    for p in percentiles_to_test_hmm_tuned:
        epsilon_candidate_optuna_hmm = np.percentile(hmm_log_likelihoods_train_normal_sorted_tuned, p)
        y_pred_candidate_optuna_hmm = (hmm_log_likelihoods_test_tuned < epsilon_candidate_optuna_hmm).astype(int)
        f1_candidate_optuna_hmm = f1_score(y_test, y_pred_candidate_optuna_hmm)

        if f1_candidate_optuna_hmm > best_f1_optuna_hmm:
            best_f1_optuna_hmm = f1_candidate_optuna_hmm

    return best_f1_optuna_hmm

# Create an Optuna study for HMM tuning
print("Starting Optuna hyperparameter tuning for HMM (all features). This may take some time...")
study_hmm_tuned = optuna.create_study(direction='maximize')
study_hmm_tuned.optimize(objective_hmm_tuned, n_trials=10) # Reduced n_trials

print("\nOptuna optimization complete for HMM (all features).")
print(f"Best trial for HMM (all features):")
print(f"  Value: {study_hmm_tuned.best_value:.4f}")
print(f"  Params: {study_hmm_tuned.best_params}")

# Store the best parameters
best_n_components_hmm = study_hmm_tuned.best_params['n_components']
best_covariance_type_hmm = study_hmm_tuned.best_params['covariance_type']
```

```python
import optuna
from hmmlearn import hmm
from sklearn.metrics import f1_score
import numpy as np
from numpy.linalg import LinAlgError

def objective_hmm_tuned_fs(trial):
    # Hyperparameters to tune for selected features
    n_components_fs = trial.suggest_int('n_components_fs', 1, 5)
    covariance_type_fs = trial.suggest_categorical('covariance_type_fs', ['diag', 'full'])

    # Convert X_train_normal_selected and X_test_selected to NumPy arrays
    X_train_normal_selected_np = X_train_normal_selected
    X_test_selected_np = X_test_selected

    # Define lengths arrays where each sample is considered a sequence of length 1
    lengths_train_normal_selected = np.ones(X_train_normal_selected_np.shape[0], dtype=int)

    # Handle cases where 'full' covariance might lead to instability/long computation
    if covariance_type_fs == 'full' and n_components_fs > 1:
        print(f"Skipping trial with n_components_fs={n_components_fs} and covariance_type_fs='full' due to potential instability and long computation.")
        return 0.0

    # Initialize HMM model with suggested hyperparameters and selected features
    # Reduced n_iter to 30 for faster execution
    hmm_model_fs = hmm.GaussianHMM(
        n_components=n_components_fs,
        covariance_type=covariance_type_fs,
        n_iter=30, # Reduced iterations for speed
        random_state=42,
        init_params="mc", # Only auto-initialize Means and Covariances
        params="mc"      # Only update Means and Covariances during .fit()
    )

    # Manually initialize startprob_ and transmat_ for robustness with single-observation sequences
    hmm_model_fs.startprob_ = np.full(n_components_fs, 1.0 / n_components_fs)
    hmm_model_fs.transmat_ = np.full((n_components_fs, n_components_fs), 1.0 / n_components_fs)

    # Fit the model to the normal training data (selected features)
    try:
        hmm_model_fs.fit(X_train_normal_selected_np, lengths=lengths_train_normal_selected)
    except (ValueError, LinAlgError, np.linalg.LinAlgError) as e:
        print(f"Trial failed due to: {type(e).__name__} - {e}. Returning 0.0 F1 score.")
        return 0.0

    # Get log-likelihoods for the test data (selected features)
    # Iterate through each sample and call .score() to get per-sample log-likelihoods
    hmm_log_likelihoods_test_tuned_fs = np.array([
        hmm_model_fs.score(X_test_selected_np[i].reshape(1, -1))
        for i in range(X_test_selected_np.shape[0])
    ])

    # Get log-likelihoods for the normal training data to determine epsilon
    # Iterate through each sample and call .score() to get per-sample log-likelihoods
    hmm_log_likelihoods_train_normal_tuned_fs = np.array([
        hmm_model_fs.score(X_train_normal_selected_np[i].reshape(1, -1))
        for i in range(X_train_normal_selected_np.shape[0])
    ])

    hmm_log_likelihoods_train_normal_sorted_tuned_fs = np.sort(hmm_log_likelihoods_train_normal_tuned_fs)

    # Optimize epsilon (threshold) to maximize F1 Score
    percentiles_to_test_hmm_tuned_fs = np.arange(0.1, 10.1, 1.0) # Reduced granularity of percentile search

    best_f1_optuna_hmm_fs = -1
    for p in percentiles_to_test_hmm_tuned_fs:
        epsilon_candidate_optuna_hmm_fs = np.percentile(hmm_log_likelihoods_train_normal_sorted_tuned_fs, p)
        y_pred_candidate_optuna_hmm_fs = (hmm_log_likelihoods_test_tuned_fs < epsilon_candidate_optuna_hmm_fs).astype(int)
        f1_candidate_optuna_hmm_fs = f1_score(y_test, y_pred_candidate_optuna_hmm_fs)

        if f1_candidate_optuna_hmm_fs > best_f1_optuna_hmm_fs:
            best_f1_optuna_hmm_fs = f1_candidate_optuna_hmm_fs

    return best_f1_optuna_hmm_fs

# Create an Optuna study for HMM tuning with selected features
print("Starting Optuna hyperparameter tuning for HMM (selected features). This may take some time...")
study_hmm_tuned_fs = optuna.create_study(direction='maximize')
study_hmm_tuned_fs.optimize(objective_hmm_tuned_fs, n_trials=10) # Reduced trials for speed

print("\nOptuna optimization complete for HMM (selected features).")
print(f"Best trial for HMM (selected features):")
print(f"  Value: {study_hmm_tuned_fs.best_value:.4f}")
print(f"  Params: {study_hmm_tuned_fs.best_params}")

# Store the best parameters for selected features
best_n_components_hmm_fs = study_hmm_tuned_fs.best_params['n_components_fs']
best_covariance_type_hmm_fs = study_hmm_tuned_fs.best_params['covariance_type_fs']
```

## Exploratory Data Analysis (EDA)

This section details the steps taken for Exploratory Data Analysis, covering data loading, initial inspection, preprocessing, and distribution analysis to prepare the dataset for anomaly detection modeling.

### 1. Data Loading and Initial Inspection

*   The training and testing datasets (`UNSW_NB15_training-set.csv` and `UNSW_NB15_testing-set.csv`) were loaded and concatenated into a single DataFrame (`combined_df`).
*   **Initial Data View:** The combined dataset has `257673` rows and `45` columns, as shown in the `combined_df.head()` output.

### 2. Column Removal

*   The `id` and `attack_cat` columns were dropped as `id` is an irrelevant identifier and `attack_cat` is a categorical representation of the target `label` column (which is numerical).
*   **DataFrame Shape After Removal:** The DataFrame was reduced to `257673` rows and `43` columns.

### 3. Label Distribution Analysis

*   The distribution of the `label` column (0: Normal, 1: Attack) in the combined dataset was analyzed.
*   **Class Distribution:** The dataset exhibits a significant class imbalance with `164673` attack instances (63.9%) and `93000` normal instances (36.1%).
*   **Visual Representation:** This imbalance is clearly depicted in the `Distribution of Labels` bar plot and the `Percentage Distribution of Labels` pie chart.

### 4. Categorical Feature Encoding

*   Categorical features (`proto`, `service`, `state`) were converted into a numerical format using one-hot encoding to make them suitable for machine learning algorithms.
*   **DataFrame Shape After Encoding:** This process expanded the DataFrame from `43` to `197` columns due to the creation of new binary columns for each category.

### 5. Numerical Feature Scaling

*   All numerical features (excluding the `label` column) were scaled using `StandardScaler` to standardize their range and prevent features with larger values from dominating the learning process.
*   **Scaled Data Example:** The `combined_df.head()` output post-scaling shows the transformed feature values, centered around zero with unit variance.

### 6. Skewness Analysis and Transformation

*   Skewness was calculated for all numerical features to identify highly skewed distributions (absolute skewness > 0.5).
*   **Highly Skewed Features:** `190` out of `196` features were identified as highly skewed.
*   **Skewness Reduction:** Yeo-Johnson Power Transformation was applied to `24` identified highly skewed numerical features (excluding binary features) to make their distributions more Gaussian-like, which can improve the performance of some models.
*   **Transformed Skewness:** The skewness values after transformation indicate a reduction in the magnitude of skewness for many features.

### 7. Data Splitting (Training and Testing)

*   The preprocessed `combined_df` was split into features (`X`) and target (`y`).
*   The data was then divided into training (80%) and testing (20%) sets using `train_test_split` with `random_state=42` for reproducibility.
*   **Dataset Shapes:** `X_train` has `206138` rows and `196` columns, `X_test` has `51535` rows and `196` columns, and their respective `y_train` and `y_test` target vectors.

### 8. Training Data Class Imbalance Check

*   The class distribution in the `y_train` set was specifically examined before resampling.
*   **Training Class Distribution:** `y_train` contains `131751` attack instances (63.9%) and `74387` normal instances (36.1%), showing a similar imbalance to the overall dataset.
*   **Visual Representation:** The `Distribution of Labels in Training Set` bar plot and pie chart confirm this imbalance.

### 9. SMOTE for Class Balancing

*   The Synthetic Minority Over-sampling Technique (SMOTE) was applied to the training data (`X_train`, `y_train`) to address the class imbalance.
*   **Balanced Training Set:** After SMOTE, both normal and attack classes in `y_train_resampled` have `131751` instances each, creating a perfectly balanced dataset.
*   **Visual Representation:** The `Distribution of Labels in Training Set After SMOTE` bar plot clearly shows the equal distribution of both classes.

## Model Performance Analysis

This section provides a detailed analysis of the performance of each anomaly detection model (MVN, Isolation Forest, Autoencoder, and HMM) across various configurations. The key metrics used for evaluation are Accuracy, Precision (for the anomaly class), Recall (for the anomaly class), and F1-score (for the anomaly class).

### 1. Multivariate Normal (MVN) Model Analysis

The MVN model was evaluated with and without feature selection, and with and without Optuna hyperparameter tuning. The model's performance was relatively consistent across most configurations when dealing with all features, achieving an F1-score around **0.84**.

*   **All Features (No Tuning):** Achieved an F1-score of **0.84**, demonstrating strong baseline performance.
*   **Selected Features (No Tuning):** Performance slightly decreased to an F1-score of **0.81**, suggesting that some of the removed features were beneficial.
*   **All Features (Optuna Tuned):** Hyperparameter tuning slightly improved the F1-score to **0.84**, indicating minor benefits from regularization optimization.
*   **Selected Features (Optuna Tuned):** The F1-score remained at **0.81**, showing that tuning didn't recover the performance lost due to feature selection.

This indicates that for the MVN model, feature selection with `SelectKBest` might remove valuable information, and the optimal performance is achieved using all features with subtle improvements from hyperparameter tuning.

### 2. Isolation Forest Model Analysis

The Isolation Forest model, an ensemble-based method, showed significant improvements with tuning and performed well with selected features.

*   **All Features (No Tuning):** Started with a lower F1-score of **0.63**, indicating that out-of-the-box parameters were not optimal.
*   **Selected Features (No Tuning):** Performance improved to an F1-score of **0.69**, suggesting that feature selection might have helped by reducing noise or irrelevant features.
*   **All Features (Optuna Tuned):** Achieved a substantial F1-score of **0.85**, demonstrating the effectiveness of hyperparameter tuning on the full feature set.
*   **Selected Features (Optuna Tuned):** The F1-score was **0.84**, showing that tuning also benefits the feature-selected model, bringing it close to the performance of the tuned model with all features.

The Isolation Forest benefits greatly from hyperparameter tuning, especially when applied to the full feature set. The selected features also yield a reasonable performance, particularly after tuning.

### 3. Autoencoder Model Analysis

The Autoencoder model proved to be highly effective, especially with tuned hyperparameters.

*   **All Features (No Tuning):** Achieved an F1-score of **0.86**, indicating robust performance even without specific tuning.
*   **Selected Features (No Tuning):** The F1-score was **0.84**, similar to the MVN model, suggesting some loss of information due to feature selection.
*   **All Features (Optuna Tuned):** Maintained a high F1-score of **0.86**, showing the stability and effectiveness of the architecture.
*   **Selected Features (Optuna Tuned):** Achieved an F1-score of **0.86**, surprisingly matching the performance of the full feature set after tuning. This suggests that with proper tuning, the selected features can be highly effective for Autoencoders.

Autoencoders consistently delivered high F1-scores, making them a strong candidate for this anomaly detection task. Tuning helped maintain performance even with a reduced feature set.

### 4. Hidden Markov Model (HMM) Analysis

The HMM, while conceptually powerful for sequential data, had lower performance compared to other models, but showed notable improvement with feature selection and tuning.

*   **All Features (No Tuning):** The F1-score was **0.72**, which is lower than other models' baseline performances.
*   **Selected Features (No Tuning):** Performance significantly dropped to an F1-score of **0.67**, indicating the importance of context provided by removed features or a less optimal feature set for HMM.
*   **All Features (Optuna Tuned):** F1-score remained at **0.72**, showing minimal impact of tuning on the full feature set.
*   **Selected Features (Optuna Tuned):** The F1-score improved substantially to **0.78**, making it the best performance for the HMM model.

While generally lower-performing than other models, the HMM benefited from feature selection combined with Optuna tuning. The `full` covariance type often led to numerical instability, emphasizing the importance of `covariance_type='diag'` and careful handling of `n_components`.
