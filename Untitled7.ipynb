````python
# =============================================================================
# CELL 1 ── INSTALLS  (run once, restart runtime if prompted)
# =============================================================================
# !pip install -q optuna torch scikit-learn --upgrade

import os, sys, warnings, traceback, time, gc, math
warnings.filterwarnings("ignore")
os.environ["PYTHONWARNINGS"] = "ignore"

import numpy  as np
import pandas as pd

from sklearn.covariance      import EllipticEnvelope
from sklearn.ensemble        import IsolationForest, RandomForestClassifier
from sklearn.preprocessing   import RobustScaler, QuantileTransformer
from sklearn.decomposition   import PCA
from sklearn.svm             import OneClassSVM
from sklearn.feature_selection import (mutual_info_classif, SelectKBest,
                                        f_classif)
from sklearn.metrics         import (precision_score, recall_score, f1_score,
                                      precision_recall_curve)

try:
    import torch, torch.nn as nn
    from torch.utils.data import DataLoader, TensorDataset
    TORCH_OK = True
    DEVICE   = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"✅  PyTorch {torch.__version__}  |  device = {DEVICE}")
except ImportError:
    TORCH_OK = False
    print("⚠️   PyTorch not found — AE will be skipped")

try:
    import optuna
    optuna.logging.set_verbosity(optuna.logging.WARNING)
    OPTUNA_OK = True
    print("✅  Optuna available")
except ImportError:
    OPTUNA_OK = False
    print("⚠️   Optuna not found — Exp3/4 use default params")

SEED = 42
np.random.seed(SEED)
pd.set_option("display.float_format", "{:.4f}".format)
pd.set_option("display.max_columns", 20)
print("✅  All imports complete\n")


# =============================================================================
# CELL 2 ── EDA  (printed summary — no plots needed in Colab script mode)
# =============================================================================

def run_eda(df: pd.DataFrame, label_col: str = "Label",
            benign_label: str = "Benign") -> None:
    sep = "=" * 70
    print(sep)
    print("  EDA — EXPLORATORY DATA ANALYSIS")
    print(sep)

    n_rows, n_cols = df.shape
    print(f"\n  Dataset shape        : {n_rows:,} rows  ×  {n_cols} columns")

    # Class distribution
    vc = df[label_col].value_counts()
    print(f"\n  Class distribution ({label_col}):")
    for cls, cnt in vc.items():
        bar = "█" * int(cnt / n_rows * 40)
        tag = "← BENIGN (train normal)" if cls == benign_label else "← ANOMALY"
        print(f"    {str(cls):<35s}  {cnt:>8,}  ({cnt/n_rows*100:5.2f}%)  {bar}  {tag}")

    # Feature types
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
    print(f"\n  Numeric features     : {len(num_cols)}")
    print(f"  Categorical features : {len(cat_cols)}  {cat_cols}")

    # Null / inf analysis
    feat_cols = [c for c in num_cols if c != label_col]
    df_feat   = df[feat_cols].replace([np.inf, -np.inf], np.nan)
    null_cnt  = df_feat.isnull().sum()
    n_null    = (null_cnt > 0).sum()
    print(f"\n  Columns with NaN/Inf : {n_null}")
    if n_null:
        top5 = null_cnt[null_cnt > 0].nlargest(5)
        for col, cnt in top5.items():
            print(f"    {col:<40s}  {cnt:,} nulls")

    # Skewness
    skew  = df_feat.skew().abs()
    hi_sk = (skew > 5).sum()
    print(f"\n  High-skew (|skew|>5) : {hi_sk} / {len(feat_cols)} features")

    # Quasi-constant
    qc = sum(
        df_feat[c].value_counts(normalize=True).iloc[0] > 0.80
        for c in feat_cols
    )
    print(f"  Quasi-constant (>80%): {qc} features  ← will be dropped")

    # Correlation estimate
    sample = df_feat.sample(min(5000, len(df_feat)), random_state=SEED)
    corr   = sample.corr().abs()
    upper  = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    hi_cor = (upper > 0.92).sum().sum()
    print(f"  Highly correlated pairs (|r|>0.92): {hi_cor}")

    # Per-class numeric summary for top-5 variance features
    top5_var = df_feat.var().nlargest(5).index.tolist()
    print(f"\n  Per-class stats for top-5 variance features:")
    print(f"  {'Feature':<35s}  {'Benign μ':>10s}  {'Benign σ':>10s}  "
          f"{'Anomaly μ':>10s}  {'Anomaly σ':>10s}")
    print(f"  {'-'*35}  {'-'*10}  {'-'*10}  {'-'*10}  {'-'*10}")
    benign_mask  = df[label_col] == benign_label
    for col in top5_var:
        bm = df.loc[benign_mask,  col].mean()
        bs = df.loc[benign_mask,  col].std()
        am = df.loc[~benign_mask, col].mean()
        as_ = df.loc[~benign_mask, col].std()
        print(f"  {col:<35s}  {bm:>10.3f}  {bs:>10.3f}  {am:>10.3f}  {as_:>10.3f}")

    print(f"\n{sep}\n")


run_eda(df, label_col="Label", benign_label="Benign")


# =============================================================================
# CELL 3 ── FEATURE ENGINEERING  (tuned thresholds — keeps more features)
# =============================================================================

def engineer_features(df: pd.DataFrame,
                       label_col: str = "Label") -> tuple:
    """
    Returns (df_engineered, feature_list).
    Conservative thresholds to preserve enough features for all models.
    """
    sep = "=" * 70
    print(sep)
    print("  FEATURE ENGINEERING")
    print(sep)

    df_c    = df.replace([np.inf, -np.inf], np.nan).copy()
    num_cols = [c for c in df_c.select_dtypes(include=[np.number]).columns
                if c != label_col]
    print(f"\n  Raw numeric features : {len(num_cols)}")

    # ── 1. Fill NaN with median ───────────────────────────────────────────────
    t = time.time()
    for col in num_cols:
        med = df_c[col].median()
        df_c[col].fillna(med, inplace=True)
    print(f"  [1/5] NaN fill (median)               done  {time.time()-t:.1f}s")

    # ── 2. Clip at 0.1% / 99.9% (very conservative — preserves distribution) ─
    t = time.time()
    for col in num_cols:
        lo = df_c[col].quantile(0.001)
        hi = df_c[col].quantile(0.999)
        df_c[col] = df_c[col].clip(lo, hi)
    print(f"  [2/5] Outlier clip (0.1–99.9%)         done  {time.time()-t:.1f}s")

    # ── 3. Log1p on skewed + domain-keyword cols ──────────────────────────────
    t = time.time()
    domain_kws = [
        'duration','bytes','packets','pkt','iat','subflow','bulk','init',
        'act','idle','max','min','std','mean','len','size','win','seg',
        'hdr','payload','flow','tot','fwd','bwd','rate','count','avg'
    ]
    skewed  = df_c[num_cols].skew().abs()
    to_log  = set(skewed[skewed > 0.75].index.tolist())
    to_log |= {c for c in num_cols if any(k in c.lower() for k in domain_kws)}
    for col in sorted(to_log):
        col_min = df_c[col].min()
        if col_min < 0:
            df_c[col] -= col_min   # shift to non-negative
        df_c[col] = np.log1p(df_c[col])
    print(f"  [3/5] Log1p transform ({len(to_log)} features)    done  {time.time()-t:.1f}s")

    # ── 4. Drop quasi-constant (>80% same value) ──────────────────────────────
    t = time.time()
    qc_drop = []
    for col in num_cols:
        top_freq = df_c[col].value_counts(normalize=True).iloc[0]
        if top_freq > 0.80:                       # raised from 0.40 → 0.80
            qc_drop.append(col)
    num_cols = [c for c in num_cols if c not in qc_drop]
    print(f"  [4/5] Quasi-constant drop ({len(qc_drop)})          done  "
          f"{time.time()-t:.1f}s  → {len(num_cols)} remain")

    # ── 5. Pearson correlation pruning (|r| > 0.92) ───────────────────────────
    t = time.time()
    sample_n = min(10_000, len(df_c))
    df_samp  = df_c[num_cols].sample(sample_n, random_state=SEED)
    corr_mat = df_samp.corr().abs()
    upper    = corr_mat.where(np.triu(np.ones(corr_mat.shape), k=1).astype(bool))
    variances= df_c[num_cols].var()
    to_drop  = set()
    for col in upper.columns:
        hi_r = upper.index[upper[col] > 0.92].tolist()
        for f in hi_r:
            if f not in to_drop and col not in to_drop:
                to_drop.add(f if variances[f] <= variances[col] else col)
    num_cols = [c for c in num_cols if c not in to_drop]
    print(f"  [5/5] Correlation drop ({len(to_drop)})              done  "
          f"{time.time()-t:.1f}s  → {len(num_cols)} remain")

    print(f"\n  ✅  Final optimized_features : {len(num_cols)}")
    print(f"  Features: {num_cols}\n")
    return df_c, num_cols


df_eng, optimized_features = engineer_features(df, label_col="Label")


# =============================================================================
# CELL 4 ── DATA SLICING  (exact sizes preserved, rate-matched val sets)
# =============================================================================

TRAIN_N = 100_000
RATE_CFG = {          # test set compositions — FIXED, never change
    "5pct"  : {"n_normal": 47_500, "n_anom":  2_500, "rate": 0.050},
    "1pct"  : {"n_normal": 49_500, "n_anom":    500, "rate": 0.010},
    "0.2pct": {"n_normal": 49_900, "n_anom":    100, "rate": 0.002},
}
VAL_CFG = {           # val sets — rate-matched to test sets (small)
    "5pct"  : {"n_normal": 4_750, "n_anom":  250},
    "1pct"  : {"n_normal": 4_950, "n_anom":   50},
    "0.2pct": {"n_normal": 4_990, "n_anom":   10},
}


def safe_sample(df_pool: pd.DataFrame, n: int, seed: int) -> pd.DataFrame:
    """Sample n rows; if pool is too small, sample with replacement."""
    replace = len(df_pool) < n
    return df_pool.sample(n=n, replace=replace, random_state=seed)


def slice_data(df, features, label_col="Label", benign_label="Benign"):
    sep = "=" * 70
    print(sep)
    print("  DATA SLICING  (rate-matched val + fixed test sets)")
    print(sep)

    benign  = df[df[label_col] == benign_label].reset_index(drop=True)
    anomaly = df[df[label_col] != benign_label].reset_index(drop=True)
    print(f"\n  Total benign  : {len(benign):,}")
    print(f"  Total anomaly : {len(anomaly):,}")

    # Shuffle once deterministically
    benign_shuf  = benign.sample(frac=1, random_state=SEED).reset_index(drop=True)
    anomaly_shuf = anomaly.sample(frac=1, random_state=SEED).reset_index(drop=True)

    # ── Train (pure benign) ───────────────────────────────────────────────────
    assert len(benign_shuf) >= TRAIN_N, \
        f"Not enough benign rows! Need {TRAIN_N}, have {len(benign_shuf)}"
    train_df   = benign_shuf.iloc[:TRAIN_N]
    remain_ben = benign_shuf.iloc[TRAIN_N:].reset_index(drop=True)
    print(f"\n  Train split   : {TRAIN_N:,} benign rows")

    # ── Val sets (rate-matched) ───────────────────────────────────────────────
    # Use non-overlapping cursor slices from remain_ben
    b_cur, a_cur = 0, 0
    val_sets, test_sets = {}, {}

    print(f"\n  Val sets (rate-matched to each test rate):")
    for key, cfg in VAL_CFG.items():
        nb, na = cfg["n_normal"], cfg["n_anom"]
        b_chunk = remain_ben.iloc[b_cur:b_cur+nb] if b_cur+nb <= len(remain_ben) \
                  else safe_sample(remain_ben, nb, SEED)
        a_chunk = anomaly_shuf.iloc[a_cur:a_cur+na] if a_cur+na <= len(anomaly_shuf) \
                  else safe_sample(anomaly_shuf, na, SEED)
        b_cur  += nb;  a_cur  += na
        combined = pd.concat([b_chunk, a_chunk]).sample(frac=1, random_state=SEED)
        y        = np.where(combined[label_col] == benign_label, 1, -1)
        val_sets[key] = (combined[features].values, y)
        rate = na / (nb + na)
        print(f"    {key:7s} : {nb:,} normal + {na:,} anomaly  "
              f"(anomaly rate = {rate*100:.1f}%)")

    # ── Test sets (EXACT sizes — never change) ────────────────────────────────
    print(f"\n  Test sets:")
    for key, cfg in RATE_CFG.items():
        nb, na = cfg["n_normal"], cfg["n_anom"]
        b_chunk = remain_ben.iloc[b_cur:b_cur+nb] if b_cur+nb <= len(remain_ben) \
                  else safe_sample(remain_ben, nb, SEED+key.__hash__())
        a_chunk = anomaly_shuf.iloc[a_cur:a_cur+na] if a_cur+na <= len(anomaly_shuf) \
                  else safe_sample(anomaly_shuf, na, SEED+key.__hash__())
        b_cur  += nb;  a_cur  += na
        combined = pd.concat([b_chunk, a_chunk]).sample(frac=1, random_state=SEED)
        y        = np.where(combined[label_col] == benign_label, 1, -1)
        test_sets[key] = (combined[features].values, y)
        print(f"    {key:7s} : {nb:,} normal + {na:,} anomaly  "
              f"(total = {nb+na:,}  rate = {cfg['rate']*100:.1f}%)")

    train_X = train_df[features].values
    print(f"\n  train_X shape : {train_X.shape}")
    print(f"  ✅  Slicing complete\n")
    return train_X, val_sets, test_sets


train_X_raw, val_sets_raw, test_sets_raw = slice_data(
    df_eng, optimized_features, label_col="Label", benign_label="Benign"
)


# =============================================================================
# CELL 5 ── SCALERS  (fit on train, apply to val+test)
# =============================================================================

print("=" * 70)
print("  FITTING SCALERS")
print("=" * 70)

# RobustScaler → IForest, AE, SVM  (IQR-based, robust to outliers)
t = time.time()
scaler_R = RobustScaler()
train_R  = scaler_R.fit_transform(train_X_raw)
print(f"\n  RobustScaler    fitted  {time.time()-t:.1f}s")

# QuantileTransformer → MVN  (makes features Gaussian-distributed)
t = time.time()
scaler_Q = QuantileTransformer(
    n_quantiles=min(2000, len(train_X_raw)),
    output_distribution="normal",
    random_state=SEED
)
train_Q  = scaler_Q.fit_transform(train_X_raw)
print(f"  QuantileScaler  fitted  {time.time()-t:.1f}s")


def apply_scaler(sets_dict, scaler):
    return {k: (scaler.transform(X), y) for k, (X, y) in sets_dict.items()}


val_R  = apply_scaler(val_sets_raw,  scaler_R)
val_Q  = apply_scaler(val_sets_raw,  scaler_Q)
test_R = apply_scaler(test_sets_raw, scaler_R)
test_Q = apply_scaler(test_sets_raw, scaler_Q)
print(f"\n  train_R : {train_R.shape}   (RobustScaler)")
print(f"  train_Q : {train_Q.shape}   (QuantileTransformer)")
print(f"  ✅  Scalers applied to all splits\n")


# =============================================================================
# CELL 6 ── PER-MODEL FEATURE SELECTION
# =============================================================================
# KEY INSIGHT from v3 analysis:
#   - Only 14 features survived → too few; we relaxed quasi-constant to 0.80
#   - Each model needs a different selection strategy
#   - SVM: PCA is only useful if dim > 15; here we just use RobustScaled directly
#         with a greedy nu search

print("=" * 70)
print("  PER-MODEL FEATURE SELECTION")
print("=" * 70)
N_FEAT = len(optimized_features)
print(f"\n  Total optimized features available : {N_FEAT}")


def _synth_labels(X, noise_scale=3.0):
    """Synthetic anomalies via Gaussian perturbation for supervised selectors."""
    rng   = np.random.RandomState(SEED)
    X_neg = X + rng.normal(0, X.std(0).clip(min=1e-8) * noise_scale, X.shape)
    Xc    = np.vstack([X, X_neg])
    yc    = np.array([0]*len(X) + [1]*len(X_neg))
    return Xc, yc


def feat_idx(names_sub, all_names):
    m = {n: i for i, n in enumerate(all_names)}
    return [m[n] for n in names_sub if n in m]


# ── MVN: F-test top-k, then greedy decorrelation ────────────────────────────
def select_mvn(X_Q, all_names, k_init=30, k_final=20):
    """
    MVN assumes multivariate Gaussian. Select high-signal, low-correlation
    features for a well-conditioned covariance matrix.
    """
    Xc, yc = _synth_labels(X_Q)
    scores = f_classif(Xc, yc)[0]
    scores = np.nan_to_num(scores, nan=0.0)
    cand_idx   = np.argsort(scores)[::-1][:min(k_init, N_FEAT)]
    cand_names = [all_names[i] for i in cand_idx]
    cand_X     = X_Q[:, cand_idx]
    corr       = np.corrcoef(cand_X.T)
    chosen     = [0]
    for i in range(1, len(cand_idx)):
        if len(chosen) >= k_final:
            break
        max_r = max(abs(corr[i, j]) for j in chosen)
        if max_r < 0.75:          # allow slightly higher r than v3 (0.70→0.75)
            chosen.append(i)
    selected = [cand_names[i] for i in chosen]
    print(f"  MVN           : {len(selected):2d} features  "
          f"(F-test top-{k_init} → greedy decorrelation to {k_final})")
    print(f"    {selected}")
    return selected


# ── IForest: RF importance — keep top-k or all if dim small ─────────────────
def select_iforest(X_R, all_names, k=30):
    if N_FEAT <= 40:
        print(f"  IsolationForest: {N_FEAT:2d} features  (all — dim ≤ 40)")
        return list(all_names)
    Xc, yc = _synth_labels(X_R)
    rf = RandomForestClassifier(n_estimators=80, max_depth=10,
                                 n_jobs=-1, random_state=SEED)
    rf.fit(Xc, yc)
    idx = np.argsort(rf.feature_importances_)[::-1][:k]
    sel = [all_names[i] for i in idx]
    print(f"  IsolationForest: {len(sel):2d} features  (RF importance top-{k})")
    print(f"    {sel[:10]}{'...' if len(sel)>10 else ''}")
    return sel


# ── AE: Mutual Information top-k ────────────────────────────────────────────
def select_ae(X_R, all_names, k=25):
    Xc, yc = _synth_labels(X_R, noise_scale=2.5)
    mi  = mutual_info_classif(Xc, yc, random_state=SEED, n_neighbors=5)
    idx = np.argsort(mi)[::-1][:min(k, N_FEAT)]
    sel = [all_names[i] for i in idx]
    print(f"  Autoencoder    : {len(sel):2d} features  "
          f"(Mutual Information top-{min(k,N_FEAT)})")
    print(f"    {sel[:10]}{'...' if len(sel)>10 else ''}")
    return sel


# ── SVM: Combined RF+MI ranking, optionally PCA-compressed ──────────────────
def select_svm(X_R, all_names, k=20):
    """
    SVM RBF kernel: combine RF and MI ranks for robust feature set.
    Use PCA only if selected dim > 15 (adds compression benefit).
    """
    Xc, yc = _synth_labels(X_R)
    rf = RandomForestClassifier(n_estimators=60, max_depth=8,
                                 n_jobs=-1, random_state=SEED)
    rf.fit(Xc, yc)
    mi  = mutual_info_classif(Xc, yc, random_state=SEED, n_neighbors=3)
    rf_ranks = rf.feature_importances_.argsort().argsort().astype(float)
    mi_ranks = mi.argsort().argsort().astype(float)
    rf_ranks /= max(rf_ranks.max(), 1e-9)
    mi_ranks /= max(mi_ranks.max(), 1e-9)
    combined = 0.5 * rf_ranks + 0.5 * mi_ranks
    idx = np.argsort(combined)[::-1][:min(k, N_FEAT)]
    sel = [all_names[i] for i in idx]
    # PCA on top of selected features (only if dim > 12)
    pca = None
    if len(sel) > 12:
        n_comp = min(12, len(sel))
        pca    = PCA(n_components=n_comp, random_state=SEED)
        pca.fit(X_R[:, feat_idx(sel, all_names)])
        var_exp = pca.explained_variance_ratio_.cumsum()[-1]
        print(f"  One-Class SVM  : {len(sel):2d} features → PCA {n_comp} comps  "
              f"({var_exp*100:.1f}% var)  (RF+MI ensemble)")
    else:
        print(f"  One-Class SVM  : {len(sel):2d} features  (RF+MI — no PCA needed)")
    print(f"    {sel[:10]}{'...' if len(sel)>10 else ''}")
    return sel, pca


t = time.time()
mvn_feats           = select_mvn(train_Q, optimized_features)
ifor_feats          = select_iforest(train_R, optimized_features)
ae_feats            = select_ae(train_R, optimized_features)
svm_feats, svm_pca  = select_svm(train_R, optimized_features)

mvn_idx  = feat_idx(mvn_feats,  optimized_features)
ifor_idx = feat_idx(ifor_feats, optimized_features)
ae_idx   = feat_idx(ae_feats,   optimized_features)
svm_idx  = feat_idx(svm_feats,  optimized_features)

print(f"\n  Feature selection total time : {time.time()-t:.1f}s")
print(f"  ✅  Per-model features ready\n")


# =============================================================================
# CELL 7 ── THRESHOLD CALIBRATION  (PR-curve + percentile fallback)
# =============================================================================

def calibrate_threshold(scores: np.ndarray, val_y: np.ndarray) -> float:
    """
    Uses precision_recall_curve over ALL natural thresholds.
    Selects threshold maximising weighted score = 0.6*F1 + 0.4*Recall.
    Weighting recall avoids P=1, R=0 degenerate solutions at low anomaly rates.
    Falls back to percentile sweep on failure.
    """
    anom_scores = -scores           # higher = more anomalous
    is_anom     = (val_y == -1).astype(int)

    try:
        if is_anom.sum() == 0:
            raise ValueError("No anomalies in val set")
        prec, rec, thrs = precision_recall_curve(is_anom, anom_scores)
        f1 = np.where((prec + rec) > 0,
                       2*prec*rec / (prec + rec + 1e-12), 0.0)
        # Weighted: favour recall slightly at very low anomaly rates
        score = 0.6 * f1 + 0.4 * rec
        best  = np.argmax(score)
        return -thrs[min(best, len(thrs)-1)]      # convert back

    except Exception:
        # Percentile fallback: scan 0.2% – 25% in 300 steps
        best_s, best_t = -1.0, np.percentile(scores, 5)
        for pct in np.linspace(0.2, 25, 300):
            t     = np.percentile(scores, pct)
            preds = np.where(scores >= t, 1, -1)
            if len(np.unique(preds)) < 2:
                continue
            f1 = f1_score(val_y, preds, pos_label=-1, zero_division=0)
            r  = recall_score(val_y, preds, pos_label=-1, zero_division=0)
            s  = 0.6*f1 + 0.4*r
            if s > best_s:
                best_s, best_t = s, t
        return best_t


def evaluate_all_rates(score_fn, val_sets, test_sets) -> dict:
    """
    For each anomaly rate: calibrate threshold on matching val set,
    then evaluate on matching test set.
    Returns {rate_key: {precision, recall, f1, threshold}}.
    """
    results = {}
    for key in ["5pct", "1pct", "0.2pct"]:
        try:
            vX, vy   = val_sets[key]
            tX, ty   = test_sets[key]
            vscores  = score_fn(vX)
            tscores  = score_fn(tX)
            if np.any(np.isnan(vscores)) or np.any(np.isnan(tscores)):
                raise ValueError("NaN in scores")
            thresh   = calibrate_threshold(vscores, vy)
            y_pred   = np.where(tscores >= thresh, 1, -1)
            results[key] = {
                "precision": precision_score(ty, y_pred, pos_label=-1, zero_division=0),
                "recall"   : recall_score   (ty, y_pred, pos_label=-1, zero_division=0),
                "f1"       : f1_score       (ty, y_pred, pos_label=-1, zero_division=0),
                "threshold": float(thresh),
            }
        except Exception as e:
            print(f"      ⚠️  Error at {key}: {e}")
            results[key] = {"precision": np.nan, "recall": np.nan,
                             "f1": np.nan, "threshold": np.nan}
    return results


# =============================================================================
# CELL 8 ── MODEL IMPLEMENTATIONS
# =============================================================================

# ── 8a. MVN via EllipticEnvelope (MCD) ───────────────────────────────────────

class MVNDetector:
    def __init__(self, contamination=0.05, support_fraction=None):
        self.contamination    = float(np.clip(contamination, 0.001, 0.499))
        self.support_fraction = support_fraction
        self._model  = None
        self._fallback = False

    def fit(self, X):
        sf = self.support_fraction
        if sf is not None:
            # sklearn requires sf > (p+1)/n and ≤ 1
            min_sf = (X.shape[1] + 1) / X.shape[0] + 0.01
            sf     = float(np.clip(sf, min_sf, 1.0))
        try:
            kw = {"contamination": self.contamination, "random_state": SEED}
            if sf is not None:
                kw["support_fraction"] = sf
            self._model = EllipticEnvelope(**kw).fit(X)
        except Exception as e:
            print(f"      EllipticEnvelope fallback ({e})")
            self._fallback = True
            self._mean = X.mean(0)
            self._std  = np.clip(X.std(0), 1e-8, None)
        return self

    def decision_function(self, X):
        if not self._fallback:
            return self._model.decision_function(X)
        return -((X - self._mean) / self._std).pow(2).sum(axis=1)


# ── 8b. IsolationForest ───────────────────────────────────────────────────────

class IForestDetector:
    def __init__(self, n_estimators=300, contamination=0.05,
                 max_samples=0.8, max_features=0.8):
        self.params = dict(
            n_estimators = int(n_estimators),
            contamination= float(np.clip(contamination, 0.001, 0.499)),
            max_samples  = float(max_samples),
            max_features = float(max_features),
            random_state = SEED, n_jobs=-1,
        )
        self._model = None

    def fit(self, X):
        self._model = IsolationForest(**self.params).fit(X)
        return self

    def decision_function(self, X):
        return self._model.decision_function(X)


# ── 8c. FAST Autoencoder  (designed for < 8 min on CPU) ─────────────────────
# Architecture: input → 128 → 64 → latent → 64 → 128 → input
# Fast by: (1) shallow depth  (2) epochs ≤ 30 on CPU / 50 on GPU
#          (3) batch_size = 1024  (4) BatchNorm for fast convergence
#          (5) early stopping patience = 5

if TORCH_OK:
    class LightAE(nn.Module):
        """Lightweight AE — 2-layer encoder/decoder with BatchNorm."""
        def __init__(self, d, h1=128, h2=64, latent=20, dropout=0.15):
            super().__init__()
            self.enc = nn.Sequential(
                nn.Linear(d, h1),  nn.BatchNorm1d(h1), nn.ReLU(),
                nn.Dropout(dropout),
                nn.Linear(h1, h2), nn.BatchNorm1d(h2), nn.ReLU(),
                nn.Linear(h2, latent),
            )
            self.dec = nn.Sequential(
                nn.Linear(latent, h2), nn.BatchNorm1d(h2), nn.ReLU(),
                nn.Linear(h2, h1),    nn.BatchNorm1d(h1), nn.ReLU(),
                nn.Dropout(dropout),
                nn.Linear(h1, d),
            )

        def forward(self, x):
            return self.dec(self.enc(x))


    class AEDetector:
        def __init__(self, h1=128, h2=64, latent=20, dropout=0.15,
                     epochs=None, lr=5e-4, batch_size=1024, patience=5):
            self.h1, self.h2, self.latent = h1, h2, latent
            self.dropout   = dropout
            # Auto-limit epochs by device to stay within 40-min budget
            max_ep         = 50 if DEVICE == "cuda" else 30
            self.epochs    = min(epochs or max_ep, max_ep)
            self.lr        = lr
            self.bs        = batch_size
            self.patience  = patience
            self._model    = None
            self._w        = None
            self._smin     = None
            self._smax     = None

        def _t(self, X):
            return torch.tensor(X.astype(np.float32)).to(DEVICE)

        def fit(self, X):
            d = X.shape[1]
            self._model = LightAE(d, self.h1, self.h2, self.latent,
                                   self.dropout).to(DEVICE)
            # Variance weights for reconstruction loss
            var      = np.clip(X.var(0), 1e-8, None)
            self._w  = torch.tensor(var / var.sum(), dtype=torch.float32).to(DEVICE)

            loader = DataLoader(
                TensorDataset(self._t(X), self._t(X)),
                batch_size=self.bs, shuffle=True, drop_last=True,
                num_workers=0, pin_memory=(DEVICE == "cuda")
            )
            opt   = torch.optim.Adam(self._model.parameters(), lr=self.lr,
                                      weight_decay=1e-5)
            sched = torch.optim.lr_scheduler.CosineAnnealingLR(
                opt, T_max=self.epochs, eta_min=1e-6
            )
            best_loss, no_imp, best_state = np.inf, 0, None
            self._model.train()

            for ep in range(self.epochs):
                ep_loss = 0.0
                for xb, _ in loader:
                    opt.zero_grad()
                    out  = self._model(xb)
                    loss = ((out - xb)**2 * self._w).mean()
                    if torch.isnan(loss):
                        raise ValueError(f"NaN loss at epoch {ep+1}")
                    loss.backward()
                    nn.utils.clip_grad_norm_(self._model.parameters(), 1.0)
                    opt.step()
                    ep_loss += loss.item()
                sched.step()
                avg = ep_loss / max(len(loader), 1)
                if avg < best_loss - 1e-8:
                    best_loss, no_imp = avg, 0
                    best_state = {k: v.cpu().clone()
                                  for k, v in self._model.state_dict().items()}
                else:
                    no_imp += 1
                if no_imp >= self.patience:
                    print(f"      Early stop epoch {ep+1}  (best={best_loss:.6f})")
                    break

            if best_state:
                self._model.load_state_dict(
                    {k: v.to(DEVICE) for k, v in best_state.items()})

            # Calibrate score range
            raw        = self._raw(X)
            self._smin = raw.min()
            self._smax = raw.max()
            return self

        def _raw(self, X):
            self._model.eval()
            chunks = []
            with torch.no_grad():
                for i in range(0, len(X), 2048):
                    xb  = self._t(X[i:i+2048])
                    err = ((self._model(xb) - xb)**2 * self._w).mean(1)
                    chunks.append(err.cpu().numpy())
            return np.concatenate(chunks)

        def decision_function(self, X):
            raw  = self._raw(X)
            raw  = np.clip(raw, self._smin, self._smax)
            norm = (raw - self._smin) / (self._smax - self._smin + 1e-12)
            return -norm          # higher = more normal


# ── 8d. One-Class SVM ────────────────────────────────────────────────────────

class SVMDetector:
    def __init__(self, nu=0.05, gamma="scale", pca=None):
        self.nu    = float(np.clip(nu, 0.001, 0.499))
        self.gamma = gamma
        self.pca   = pca
        self._model = None

    def _transform(self, X):
        return self.pca.transform(X) if self.pca is not None else X

    def fit(self, X):
        Xt  = self._transform(X)
        # Sub-sample for speed: OC-SVM is O(n^2)
        n   = min(len(Xt), 30_000)
        idx = np.random.choice(len(Xt), n, replace=False)
        self._model = OneClassSVM(nu=self.nu, kernel="rbf",
                                   gamma=self.gamma).fit(Xt[idx])
        return self

    def decision_function(self, X):
        return self._model.decision_function(self._transform(X))


# =============================================================================
# CELL 9 ── OPTUNA OBJECTIVES  (25 trials, weighted avg F1 across 3 val rates)
# =============================================================================

N_TRIALS = 25

def _avg_f1(score_fn, val_sets_sliced):
    """Weighted average F1 across the 3 rate-matched val sets."""
    weights = {"5pct": 0.50, "1pct": 0.30, "0.2pct": 0.20}
    total   = 0.0
    for key, w in weights.items():
        try:
            vX, vy = val_sets_sliced[key]
            sc     = score_fn(vX)
            if np.any(np.isnan(sc)):
                continue
            thr    = calibrate_threshold(sc, vy)
            pred   = np.where(sc >= thr, 1, -1)
            f1     = f1_score(vy, pred, pos_label=-1, zero_division=0)
            r      = recall_score(vy, pred, pos_label=-1, zero_division=0)
            total += w * (0.6*f1 + 0.4*r)
        except Exception:
            pass
    return total


def _slice_sets(sets_dict, idx):
    return {k: (X[:, idx], y) for k, (X, y) in sets_dict.items()}


def make_obj_mvn(X_tr, val_Q_sliced):
    def obj(trial):
        c  = trial.suggest_float("contamination",    0.005, 0.30)
        sf = trial.suggest_float("support_fraction", 0.55,  0.99)
        try:
            m = MVNDetector(contamination=c, support_fraction=sf).fit(X_tr)
            return _avg_f1(m.decision_function, val_Q_sliced)
        except Exception:
            return 0.0
    return obj


def make_obj_iforest(X_tr, val_R_sliced):
    def obj(trial):
        kw = dict(
            n_estimators = trial.suggest_int  ("n_estimators", 100,  600),
            contamination= trial.suggest_float("contamination", 0.005, 0.25),
            max_samples  = trial.suggest_float("max_samples",   0.5,  1.0),
            max_features = trial.suggest_float("max_features",  0.4,  1.0),
        )
        try:
            m = IForestDetector(**kw).fit(X_tr)
            return _avg_f1(m.decision_function, val_R_sliced)
        except Exception:
            return 0.0
    return obj


def make_obj_ae(X_tr, val_R_sliced):
    def obj(trial):
        kw = dict(
            h1      = trial.suggest_categorical("h1",      [64, 128, 256]),
            h2      = trial.suggest_categorical("h2",      [32, 64,  128]),
            latent  = trial.suggest_categorical("latent",  [10, 16,  24 ]),
            dropout = trial.suggest_float       ("dropout", 0.05, 0.30),
            lr      = trial.suggest_float       ("lr",      1e-4, 5e-3, log=True),
            # Fix epochs to fast value during search; final train uses more
            epochs  = 15,
            batch_size = trial.suggest_categorical("batch_size", [512, 1024]),
        )
        try:
            m = AEDetector(**kw).fit(X_tr)
            return _avg_f1(m.decision_function, val_R_sliced)
        except Exception:
            return 0.0
    return obj


def make_obj_svm(X_tr, val_R_sliced, pca):
    def obj(trial):
        nu    = trial.suggest_float("nu",    0.005, 0.30, log=True)
        gamma = trial.suggest_categorical("gamma", ["scale", "auto"])
        try:
            m = SVMDetector(nu=nu, gamma=gamma, pca=pca).fit(X_tr)
            return _avg_f1(m.decision_function, val_R_sliced)
        except Exception:
            return 0.0
    return obj


def run_study(objective_fn, n_trials=N_TRIALS):
    study = optuna.create_study(
        direction="maximize",
        sampler=optuna.samplers.TPESampler(seed=SEED, n_startup_trials=8),
        pruner =optuna.pruners.MedianPruner(n_startup_trials=5,
                                             n_warmup_steps=3),
    )
    study.optimize(objective_fn, n_trials=n_trials,
                   show_progress_bar=False, catch=(Exception,))
    return study.best_params, study.best_value


# =============================================================================
# CELL 10 ── EXECUTION MATRIX  (4 × 4 loop)
# =============================================================================

def print_model_header(model_name, dim, use_optuna):
    print(f"\n  ▶  {model_name}")
    print(f"     dim={dim}  |  Optuna={'YES' if use_optuna else 'NO'}")


def print_rate_result(key, p, r, f, thr):
    flag = ("🟢" if f >= 0.75 else "🟡" if f >= 0.50 else "🔴")
    print(f"     {flag}  [{key:7s}]  "
          f"Prec={p:.4f}  Rec={r:.4f}  F1={f:.4f}  "
          f"thr={thr:.5f}")


def run_matrix():
    sep = "=" * 70
    print(f"\n{sep}")
    print("  EXECUTION MATRIX  —  4 models × 4 experiments × 3 test rates")
    print(sep)

    nan_row = {k: {"precision": np.nan, "recall": np.nan,
                    "f1": np.nan, "threshold": np.nan}
               for k in ["5pct", "1pct", "0.2pct"]}

    exp_configs = [
        ("all",   False, "Exp1: All Feats / No Tuning"),
        ("top",   False, "Exp2: Selected Feats / No Tuning"),
        ("all",   True,  "Exp3: All Feats / Optuna"),
        ("top",   True,  "Exp4: Selected Feats / Optuna"),
    ]

    all_results = {}
    TOTAL_T0    = time.time()

    for feat_key, use_optuna, exp_label in exp_configs:
        is_all = (feat_key == "all")
        print(f"\n{'─'*70}")
        print(f"  {exp_label}")
        print(f"{'─'*70}")

        # ── Feature slices for this experiment ────────────────────────────────
        # MVN uses Quantile-scaled data
        mvn_tr  = train_Q if is_all else train_Q[:, mvn_idx]
        mvn_v   = val_Q   if is_all else _slice_sets(val_Q,  mvn_idx)
        mvn_t   = test_Q  if is_all else _slice_sets(test_Q, mvn_idx)
        mvn_v   = mvn_v   if isinstance(mvn_v, dict) else val_Q
        mvn_t   = mvn_t   if isinstance(mvn_t, dict) else test_Q

        # IForest uses Robust-scaled data
        ifi_idx_use = list(range(N_FEAT)) if is_all else ifor_idx
        ifi_tr  = train_R[:, ifi_idx_use]
        ifi_v   = _slice_sets(val_R,  ifi_idx_use)
        ifi_t   = _slice_sets(test_R, ifi_idx_use)

        # AE uses Robust-scaled data
        ae_idx_use = list(range(N_FEAT)) if is_all else ae_idx
        ae_tr   = train_R[:, ae_idx_use]
        ae_v    = _slice_sets(val_R,  ae_idx_use)
        ae_t    = _slice_sets(test_R, ae_idx_use)

        # SVM uses Robust-scaled + optional PCA
        svm_idx_use = list(range(N_FEAT)) if is_all else svm_idx
        svm_pca_use = None if is_all else svm_pca
        svm_tr  = train_R[:, svm_idx_use]
        svm_v   = _slice_sets(val_R,  svm_idx_use)
        svm_t   = _slice_sets(test_R, svm_idx_use)

        # ── MVN ───────────────────────────────────────────────────────────────
        print_model_header("MVN (EllipticEnvelope / MCD)",
                           mvn_tr.shape[1], use_optuna)
        t0 = time.time()
        try:
            if use_optuna and OPTUNA_OK:
                bp, bv = run_study(make_obj_mvn(mvn_tr, mvn_v))
                print(f"     Optuna → best_params={bp}  val_score={bv:.4f}")
                m = MVNDetector(**bp).fit(mvn_tr)
            else:
                m = MVNDetector(contamination=0.05).fit(mvn_tr)
            res = evaluate_all_rates(m.decision_function, mvn_v, mvn_t)
            for k, v in res.items():
                print_rate_result(k, v["precision"], v["recall"],
                                   v["f1"], v["threshold"])
        except Exception as e:
            print(f"     ⚠️  ERROR: {e}\n     {traceback.format_exc()[-200:]}")
            res = nan_row.copy()
        all_results[("MVN (MCD)", exp_label)] = res
        print(f"     ⏱  {time.time()-t0:.1f}s")
        gc.collect()

        # ── IsolationForest ───────────────────────────────────────────────────
        print_model_header("IsolationForest", ifi_tr.shape[1], use_optuna)
        t0 = time.time()
        try:
            if use_optuna and OPTUNA_OK:
                bp, bv = run_study(make_obj_iforest(ifi_tr, ifi_v))
                print(f"     Optuna → best_params={bp}  val_score={bv:.4f}")
                m = IForestDetector(**bp).fit(ifi_tr)
            else:
                m = IForestDetector(n_estimators=300, contamination=0.05).fit(ifi_tr)
            res = evaluate_all_rates(m.decision_function, ifi_v, ifi_t)
            for k, v in res.items():
                print_rate_result(k, v["precision"], v["recall"],
                                   v["f1"], v["threshold"])
        except Exception as e:
            print(f"     ⚠️  ERROR: {e}")
            res = nan_row.copy()
        all_results[("IsolationForest", exp_label)] = res
        print(f"     ⏱  {time.time()-t0:.1f}s")
        gc.collect()

        # ── Autoencoder ───────────────────────────────────────────────────────
        if TORCH_OK:
            print_model_header("Autoencoder (LightAE)", ae_tr.shape[1], use_optuna)
            t0 = time.time()
            try:
                if use_optuna and OPTUNA_OK:
                    bp, bv = run_study(make_obj_ae(ae_tr, ae_v), n_trials=15)
                    print(f"     Optuna → best_params={bp}  val_score={bv:.4f}")
                    # Final training with more epochs using best arch
                    final_kw           = dict(bp)
                    max_ep_final       = 50 if DEVICE == "cuda" else 30
                    final_kw["epochs"] = max_ep_final
                    m = AEDetector(**final_kw).fit(ae_tr)
                else:
                    m = AEDetector(h1=128, h2=64, latent=20,
                                    epochs=None, lr=5e-4).fit(ae_tr)
                res = evaluate_all_rates(m.decision_function, ae_v, ae_t)
                for k, v in res.items():
                    print_rate_result(k, v["precision"], v["recall"],
                                       v["f1"], v["threshold"])
            except Exception as e:
                print(f"     ⚠️  ERROR: {e}\n     {traceback.format_exc()[-300:]}")
                res = nan_row.copy()
            all_results[("Autoencoder", exp_label)] = res
            print(f"     ⏱  {time.time()-t0:.1f}s")
            gc.collect()
            if DEVICE == "cuda":
                torch.cuda.empty_cache()

        # ── One-Class SVM ─────────────────────────────────────────────────────
        print_model_header("One-Class SVM", svm_tr.shape[1], use_optuna)
        t0 = time.time()
        try:
            if use_optuna and OPTUNA_OK:
                bp, bv = run_study(make_obj_svm(svm_tr, svm_v, svm_pca_use))
                print(f"     Optuna → best_params={bp}  val_score={bv:.4f}")
                m = SVMDetector(**bp, pca=svm_pca_use).fit(svm_tr)
            else:
                m = SVMDetector(nu=0.05, pca=svm_pca_use).fit(svm_tr)
            res = evaluate_all_rates(m.decision_function, svm_v, svm_t)
            for k, v in res.items():
                print_rate_result(k, v["precision"], v["recall"],
                                   v["f1"], v["threshold"])
        except Exception as e:
            print(f"     ⚠️  ERROR: {e}")
            res = nan_row.copy()
        all_results[("One-Class SVM", exp_label)] = res
        print(f"     ⏱  {time.time()-t0:.1f}s")
        gc.collect()

    elapsed = time.time() - TOTAL_T0
    print(f"\n{'='*70}")
    print(f"  ✅  MATRIX COMPLETE  |  Total time: {elapsed/60:.1f} min")
    print(f"{'='*70}\n")
    return all_results


all_results = run_matrix()


# =============================================================================
# CELL 11 ── RESULTS TABLE  (multi-index DataFrame)
# =============================================================================

def build_table(all_results):
    rate_map    = {"5pct": "5% Anomaly", "1pct": "1% Anomaly",
                   "0.2pct": "0.2% Anomaly"}
    col_tuples  = [(rl, m) for rl in rate_map.values()
                   for m  in ["Precision", "Recall", "F1"]]
    col_idx     = pd.MultiIndex.from_tuples(
                      col_tuples, names=["Test Rate", "Metric"])
    model_order = ["MVN (MCD)", "IsolationForest", "Autoencoder", "One-Class SVM"]
    exp_order   = [
        "Exp1: All Feats / No Tuning",
        "Exp2: Selected Feats / No Tuning",
        "Exp3: All Feats / Optuna",
        "Exp4: Selected Feats / Optuna",
    ]
    rows, row_idx = [], []
    for m in model_order:
        for e in exp_order:
            key = (m, e)
            if key not in all_results:
                continue
            r   = all_results[key]
            row = []
            for rk in ["5pct", "1pct", "0.2pct"]:
                rv = r.get(rk, {})
                row.extend([rv.get("precision", np.nan),
                             rv.get("recall",    np.nan),
                             rv.get("f1",        np.nan)])
            rows.append(row)
            row_idx.append((m, e))
    mi = pd.MultiIndex.from_tuples(row_idx, names=["Model", "Experiment"])
    return pd.DataFrame(rows, index=mi, columns=col_idx)


results_df = build_table(all_results)


# =============================================================================
# CELL 12 ── DISPLAY  (styled table + plain text fallback)
# =============================================================================

def _color_f1(v):
    if pd.isna(v):     return "background:#e0e0e0;color:#888"
    if v >= 0.85:      return "background:#1b5e20;color:#fff;font-weight:500"
    if v >= 0.75:      return "background:#388e3c;color:#fff;font-weight:500"
    if v >= 0.65:      return "background:#66bb6a;color:#1b5e20;font-weight:500"
    if v >= 0.50:      return "background:#fff9c4;color:#6b5000"
    if v >= 0.35:      return "background:#ffe0b2;color:#7c4000"
    return                    "background:#ffcdd2;color:#b71c1c"

print("=" * 70)
print("  FINAL BENCHMARK RESULTS — Anomaly Class (-1) Metrics")
print("=" * 70)

try:
    from IPython.display import display
    f1_cols = [c for c in results_df.columns if c[1] == "F1"]
    styled  = (
        results_df.style
            .format("{:.4f}", na_rep="—")
            .applymap(_color_f1, subset=f1_cols)
            .set_caption(
                "Anomaly Detection Benchmark v4  |  Anomaly Class (−1)  |  "
                "🟩≥0.85  🟢≥0.75  🍃≥0.65  🟡≥0.50  🟠≥0.35  🔴<0.35"
            )
            .set_table_styles([
                {"selector": "caption",
                 "props": [("font-size","14px"),("font-weight","500"),
                           ("text-align","left"),("padding","8px 0")]},
                {"selector": "th",
                 "props": [("background","#1c2833"),("color","#fff"),
                           ("text-align","center"),("padding","8px 12px")]},
                {"selector": "td",
                 "props": [("text-align","center"),("padding","6px 14px"),
                           ("border","1px solid #ddd")]},
            ])
    )
    display(styled)
except Exception:
    print(results_df.to_string())


# ── Plain-text per-model summary ─────────────────────────────────────────────
print("\n" + "=" * 70)
print("  DETAILED RESULTS BY MODEL")
print("=" * 70)

for model in ["MVN (MCD)", "IsolationForest", "Autoencoder", "One-Class SVM"]:
    print(f"\n  ── {model} ──")
    header = f"  {'Experiment':<35s}  " \
             f"{'P@5%':>6} {'R@5%':>6} {'F1@5%':>6}  " \
             f"{'P@1%':>6} {'R@1%':>6} {'F1@1%':>6}  " \
             f"{'P@0.2%':>6} {'R@0.2%':>6} {'F1@0.2%':>7}"
    print(header)
    print("  " + "-" * (len(header)-2))
    exp_labels = [
        "Exp1: All Feats / No Tuning",
        "Exp2: Selected Feats / No Tuning",
        "Exp3: All Feats / Optuna",
        "Exp4: Selected Feats / Optuna",
    ]
    for exp in exp_labels:
        key = (model, exp)
        if key not in all_results:
            continue
        r    = all_results[key]
        exp_short = exp.split(":")[0]
        vals = []
        for rk in ["5pct", "1pct", "0.2pct"]:
            rv = r.get(rk, {})
            p  = rv.get("precision", np.nan)
            rc = rv.get("recall",    np.nan)
            f1 = rv.get("f1",        np.nan)
            vals.extend([p, rc, f1])
        row = f"  {exp_short:<35s}  "
        row += "  ".join(
            f"{v:6.4f}" if not np.isnan(v) else f"{'—':>6}" for v in vals
        )
        print(row)


# =============================================================================
# CELL 13 ── CHAMPION ANALYSIS  + EXPORT
# =============================================================================

print("\n" + "=" * 70)
print("  CHAMPION ANALYSIS")
print("=" * 70)

try:
    f1_5   = results_df[("5% Anomaly",   "F1")]
    f1_1   = results_df[("1% Anomaly",   "F1")]
    f1_02  = results_df[("0.2% Anomaly", "F1")]

    # Composite: weighted average (higher weight to harder rates)
    composite = (0.40 * f1_5 + 0.35 * f1_1 + 0.25 * f1_02).dropna()
    composite = composite.sort_values(ascending=False)

    medals = ["🥇", "🥈", "🥉"] + ["  "] * 50
    print(f"\n  Ranking (composite = 0.40×F1@5% + 0.35×F1@1% + 0.25×F1@0.2%)\n")
    print(f"  {'#':>3}  {'Model':<22}  {'Experiment':<37}  "
          f"{'Comp':>6}  {'F1@5%':>6}  {'F1@1%':>6}  {'F1@0.2%':>7}")
    print(f"  {'-'*3}  {'-'*22}  {'-'*37}  {'-'*6}  {'-'*6}  {'-'*6}  {'-'*7}")
    for i, (k, v) in enumerate(composite.items(), 1):
        print(f"  {medals[i-1]} {i:>2}  {k[0]:<22}  {k[1]:<37}  "
              f"{v:6.4f}  {f1_5.get(k,np.nan):6.4f}  "
              f"{f1_1.get(k,np.nan):6.4f}  {f1_02.get(k,np.nan):7.4f}")

    champ = composite.idxmax()
    print(f"\n  {'─'*70}")
    print(f"  🏆  CHAMPION : {champ[0]}  |  {champ[1]}")
    print(f"       F1 @ 5%  anomaly rate  : {f1_5[champ]:.4f}")
    print(f"       F1 @ 1%  anomaly rate  : {f1_1[champ]:.4f}")
    print(f"       F1 @ 0.2% anomaly rate : {f1_02[champ]:.4f}")
    print(f"       Composite score         : {composite[champ]:.4f}")

    # Per-rate champions
    print(f"\n  Per-rate champions:")
    for label, col in [("5%", f1_5), ("1%", f1_1), ("0.2%", f1_02)]:
        best = col.idxmax()
        print(f"    Best @ {label:5s} : {best[0]:<22}  {best[1]}  "
              f"F1={col[best]:.4f}")

except Exception as e:
    print(f"Champion analysis failed: {e}")
    traceback.print_exc()

# ── Export ────────────────────────────────────────────────────────────────────
try:
    out = "anomaly_benchmark_v4.csv"
    results_df.to_csv(out)
    print(f"\n  💾  Saved → {out}")
except Exception as e:
    print(f"  Export failed: {e}")

print(f"\n{'='*70}")
print(f"  ✅  PIPELINE v4 COMPLETE")
print(f"{'='*70}")
````


# Anomaly Detection Benchmark Report

This report summarizes the findings from an anomaly detection benchmark using various models and feature engineering strategies.

## 1. Dataset and Preprocessing

The dataset analyzed is `02-15-2018.csv`, containing `1,048,575` rows and `80` columns. Key characteristics include:
- **Class Distribution**: Predominantly benign traffic (`94.99%`) with a small percentage of anomalies (e.g., DoS attacks).
- **Feature Types**: `78` numeric features and `2` categorical features (`Timestamp`, `Label`).
- **Data Issues**: Identified `2` columns with NaN/Inf values (`Flow Byts/s`, `Flow Pkts/s`), `33` highly skewed features, `24` quasi-constant features, and `60` highly correlated feature pairs.

## 2. Feature Engineering

The following steps were performed to prepare the data:
1.  **NaN Imputation**: Missing values were filled with the median.
2.  **Outlier Clipping**: Values were clipped at the 0.1% and 99.9% quantiles to handle extreme outliers conservatively.
3.  **Log Transformation**: Log1p transformation was applied to `26` skewed features and domain-keyword related columns.
4.  **Quasi-Constant Feature Removal**: `24` features with >80% identical values were dropped.
5.  **Correlation Pruning**: Highly correlated feature pairs (Pearson |r| > 0.92) were reduced, resulting in `26` optimized features.

## 3. Data Slicing

The dataset was split into:
-   **Training Set**: `100,000` pure benign samples.
-   **Validation Sets**: Rate-matched to test sets for hyperparameter tuning (`5%`, `1%`, `0.2%` anomaly rates).
-   **Test Sets**: Fixed test sets with specific anomaly rates (`5%`, `1%`, `0.2%`).

## 4. Scalers

Two types of scalers were fitted on the training data and applied to all splits:
-   **RobustScaler**: Used for Isolation Forest, Autoencoder, and One-Class SVM. (Shape: `(100000, 26)`)
-   **QuantileTransformer**: Used for MVN. (Shape: `(100000, 26)`)

## 5. Per-Model Feature Selection

Each model employed a tailored feature selection strategy based on the `26` optimized features:
-   **MVN**: F-test top-30, then greedy decorrelation to `11` features.
-   **IsolationForest**: Random Forest importance top-`26` features (all features chosen as `dim <= 40`).
-   **Autoencoder**: Mutual Information top-`25` features.
-   **One-Class SVM**: Combined RF+MI ranking top-`20` features. PCA was not applied as the dimension was below the threshold for compression benefits.

## 6. Models and Benchmarking

The following anomaly detection models were benchmarked across four experiments (All Feats / No Tuning, Selected Feats / No Tuning, All Feats / Optuna, Selected Feats / Optuna):
-   **MVN (EllipticEnvelope / MCD)**
-   **IsolationForest**
-   **Autoencoder (LightAE)**
-   **One-Class SVM**

Thresholds were calibrated using precision-recall curves on validation sets, prioritizing a weighted score of `0.6*F1 + 0.4*Recall`.

## 7. Results and Champion Analysis

The models were evaluated across three anomaly rates: 5%, 1%, and 0.2%. A composite score (weighted average of F1 scores across rates) was used for overall ranking.
